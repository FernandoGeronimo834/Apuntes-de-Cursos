# Git y GitHub


### Que puedo hacer:
   * Navegar entre cualquier punto de versiones
   * Ver que dia, o que hora, o que minuto se modifico el codigo o quien lo hizo y que linea de codigo de modifico
   * Revertir cambios como volver a versiones anteriores
   * Puede trabajar con repositorios locales y remotos
   * Los repositorios son tan solo un computador que administra mi codigo, es decir lo guardare dentro de un computador y sera accesible por otros desarrolladores (a mi eleccion)

### Conceptos Basicos de Git
#### Tiene 3 estados:

* **Working directori**: Donde voy a trabajar con todos mis archivos, cuando este por crear la primera version de mi programa, lo guardare en Staging area.
* **Staging area o Staged (steish)**: Donde voy a esta agregando todos los archivos que vas a preparar para el guardado en git, donde voy a tener un listado donde estaran las diferentes versiones (version 1, version 2, etc...).
* **repository**: Cuando ya estoy decidido a poder guardar los cambios los voy a pasar al repositorio

## Comando Basicos:

   * **git init**: Manera en la que podremos decir a Git o al sistema operativo que vamos a crear un proyecto nuevo, o si tengo ya un proyecto iniciado, que a dicho proyecto voy a empezar a usar git.
   * **git add < file>**: Es para poder pasa los archivos del **Working directori** al **Staging area**.
   * **git add .**: Sube todos los archivos que al Staging area
   * **git reset .**: Si quiero volver todos los archivos de staging area al working area.
   * **git status**: Es para ver en que estado estan mis archivos, es decir saber si estan en el **Working directori** o en el **Staging area**.
   * **git status -s**: me mostrara el estado de los archivos por ej: M rojo: el archivo fue modificado. M (verde): El archivo esta en Staged listo para el commit, o ??: Si es un archivo nuevo, el cual debo subir al staged con git add.
   * **git commit**: Es para pasar mis archivos del Staging Area al Repository, es para crear mi primer foto del codigo.
   * **git commit -m "descripcion de la nueva version"**: Para evitar ir a la otra interface, y hacer todo en una sola linea de codigo 
   * **git pull**: En caso de estar trabajando con varios desarrolladores, te traes los cambios que han hecho los otros desarrolladores.
   * **git clone**: Hace un copia desde el servidor central donde esta el codigo, a tu computadora para que empiezes a trabajar.
   * **git log**: Para ver todos los commits que hemos creado (id e de los commits, usuarios, correos, fecha y hora en que se modifico). --> con la letra q me salgo si espera algun comando
   * **git log --oneline**: Solo vere las id de los commits y la descripcion del commit, solo esa informacion --> para salir de esa pantalla debo presionar la letra "q"
   * **git commit --amend**: Para editar la descripcion del commit, para editar presiono la letra i y modifico el mensaje, luego de modificar le doy a ESC y escribo el mensaje :wq! y luego enter
   * **git diff**: Para ver las diferencias hechas en los archivos. (- en rojo) las lineas eliminadas (+ verde) las lineas agregadas
   * **git branch**: Para ver el nombre de la rama en la que estoy trabajando o de las otras versionaes () la de color verde es la rama donde estoy trabajando.
   * **git checkout nombre_de_la_version**: Si quiero moverme dentro de una version
   * **git checkout -- .**: Si modifique algo y no puedo volver atras o elimine ese archivo, con este comando es que todos los archivos que se encuentren en mi repositorio y le estemos dando reconocimiento, van a ser reconstruidos a como se encontraban en el ultimo commit, eso inclusive si hay archivos que elimine.
   * **git checkout -b nombre_de_nueva_rama**: Para crear una nueva rama y entrar directamente a trabajar en esa rama (verifico con git branch)
   * **git merge rama_que_quiero_unir_a_la_master**: (Debo estar en la rama master) (unir una rama a la principal) uso este comando si quiero unir la una rama creada y a la rama master, con todos los archivos que contenia la rama creada a partir de master, para que si he creado un archivo en la nueva rama y la necesito en master, usando este comando tambien pueda tener esos archivos. Al dar enter me da para poner el nombre de la union (commit) le doy el nombre, luego :wq! y enter
   * **git branch -d nombre_de_la_rama**: Eliminar una rama luego de que ya no la necesite
   * **git commit -am "descripcion_del_commit"**: (Es una union de git add y el git commit) (Solo lo puedo usar cuando el archivo ya tiene seguimiento). lo uso para agregar el archivo y para hacer el commit. luego git push para subirlo a git hub
   * **git push**: Es para subirlo a un repositorio remoto, por ej: cuando tiene tu codigo, ya lo has trabajado, lo vas a subir a un servidor para que pueda ser accedido otros desarrolladores que tengan permisos para modificar el codigo.
