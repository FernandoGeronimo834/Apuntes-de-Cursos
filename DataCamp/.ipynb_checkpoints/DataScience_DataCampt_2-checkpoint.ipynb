## ¿Por qué son necesarias las historias?

Las personas que trabajan con datos pasan mucho tiempo recopilando y procesando datos, y modelando y evaluando resultados.

Pero, ¿qué pasa con la comunicación de ideas?

Incluso los mejores resultados podrían no tener un impacto se las ideas no se presentan de manera adecuada.

Uno de los principales desafíos es convencer a las parte interesadas adversas al cambio de que se deben hacer algunas cosas diferentemente, especialemte si estas partes interesadas no son técnicas y no se sienten cómodos con las matemáticas o las estadísticas.

<img src=".\img_2\2.png" width="500px" height="300px">


## ¿Qué es la narración de datos?

Por eso, contar historias con datos es primordial. Entonces, que es la narración de datos?

Es un mecanismo poderoso para compartir información respaldada por una narrativa convincente y visualizaciones eficientes.

**¿Por qué es tan poderoso?**

Las anécdotas impulsan la imaginación de las personas y las historias son más memorables que las métricas y agregan valor al dar contexto a los datos. Gracias a eso, la audiencia puede entender ideas. Así se capta su atención.

Esos conocimientos facilitan el proceso de toma de decisiones y, a su vez, impulsar acciones y cambios.



## Narración de Datos  (Data Storytelling)

Comencemos con la construcción de una historia de datos efectiva.

Dos conceptos nos ayudan a construir su núcleo.

El primer concepto, la historia de tres minutos, nos hace pensar: si solo teníamos tres minutos para contar una historia, ¿en qué nos enfocaríamos?

El segundo concepto, la gran idea, nos empuja a expresar el punto de vista único de nuestra historia en una oración.

Estos conceptos nos ayudan a articular nuestra historia de manera clara y concisa.


Hay tres elementos centrales para cualquier historia de datos:

Estos elementos hacen la historia 
- **Perspicaz** --> ya que deriva aprendizajes claros de nuestro análisis. 
- **Explicativos** --> porque ayudamos a nuestra audiencia a comprender sus ideas.
- **conciso** --> dando sólo hechos concretos y especificos.

Simplificaremos esto en técnico y no técnico para una mejor comprensión.

El conocimiento técnico es un continuo y depende de usted evaluar el conocimiento técnico de su audiencia.

El director de análisi probablemenete sea más técnico que el generalista de operaciones de personas.

En general, a los profesionales de datos les gusta hablar sobre sus métodos. Las audiencias menos técnicas, sin embargo, normalmente solo se preocupan por cuál es el resultado y cuáles son las implicaciones.


## ¿Cómo de técnico?

Por ejemplo, es posible que tengamos que explicar por qué las predicciones de baja precisión afectan el resultado comercial para los especialistas de la cadena de suministro y cómo podría ayudar una mejor entrada de datos por su parte.

No les importará nuestro modelo o el ajuste de hiperparámetros. Les importará que si se mejora la entrada de datos y se toma en serio, las predicciones mejorarán, tendrán que lidiar con muchas menos devoluciones y problemas de envío.


## Traducir los resultados técnicos en historias


Esto requiere traducir los resultado a historia fáciles de entender que pueden involucrar a una audiencia no técnica.

Las parte interesadas entederán nuestro resultados, contando historias con simples términos pueden facilitar los procesis de toma de decisiones y, finalmente, impulsar el cambio.

Existen diferentes estrategias para comunicar nuestros conocimientos y resulados de manera efectiva. 


## Conciencia

Lo primero es condiderar a nuestra audiencia. Cuál es su transfondo?

Comprender a nuestra audiencia nos ayuda a ajustar el contenido de nuestra historia a cada parte interesada. Entonces, enl ugar de explicar cómo funciona nuestro modelo, deben comprender el impacto y las limitaciones de nuestras predicciones.

**¿Cuánto necesitan entender para alcanzar nuestra meta?**

Debemos explicar los resultados con entusiasmo y estar atentos a cualquier duda que tengan. Explicar por qué elegimos las variables para predecir no es útil para una persona no tecnológica, debemos explicar el contexto en el que funciona nuestro modelo.

**¿Qué información necesitan?**

Nuestra historia debe ayudar a la audiencia a comprender de qué se trata nuestro proyecto.

Enumerar los coeficientes de correlación podría ser agobiante, es mejor explicar la interacción entre los rasgos del cliente.

## Jerga técnica

También tenemos que prestar atención a nuestra elección de palabras. Las siglas, por ejemplo, deben usarse con precaución. Ayudan a la comunicación si todos los entienden, pero de los contrario la perjudican. En Space X, por ejemplo, las siglas deben ser aprobadas por Musk.

## Centrarse en el impacto

Incluso si estamos interesados en mostrar los detalles técnicos, una historia persuasiva generalmente se enfoca en el impacto más que en el proceso.

Si estamos hablando de adoptar nuevas tecnologías, deberíamos centrarnos en el impacto que tiene la tecnología en el tiempo y los entregables del equipo técnico. O si estamos explicando cómo nuestro modelo predice un resultado, debemos centrarnos en los beneficios, como los beneficios financieros.


## Humildad

Debemos ser conscientes de que no todos los detalles técnicos se pueden eliminar. En este caso, debemos hacer que nuestra audiencia se sienta cómoda demostrando receptividad a las preguntas, preguntando proactivamente si está claro o si hay alguna pregunta después de una explicación y estar preparado para volver a explicar conceptos utilizando diferentes estrategias.

# Impacto en el proceso de toma de decisiones

## Narración convincente

Una vez qye hayamos encontrado información relevante de nuestros datos, tenemos que encontrar una manera significativa de presentarlos a nuestro público objetivo, usando una narrativa que incluye solo puntos clave  y nos ayuda a impulsar el cambio.

En otras palabras, necesitamos una narrativa convincente o una descripción de eventos conectados que organiza la información para involucrar a la audiencia y hacer que se interesen por los resultados o la información compartida. Esto puede sonar desafiante al principio, pero seguir una estructura narrativa ayudará.


## Estructura Narrativa


Para comenzar nuestra historia de datos, debemos mencionar los antecedentes:

   - Detalles sobre qué problemas motivó el análisi.
   - Lo que cambió en la situación previa que ameritaba un análisis.
   - Y en quién se centra el análisis.
       - ¿Estamos analizando los datos de los clientes?, De los empleados?. o de algo más?
       
       
En nuestro caso, mencionamos que frente a las expectativas de la gerencia las utilidades totales de la compañia han disminuido en los últimos tres trimestres.


Después de que se introdujo el problema, debe proporcionar evidencia de los factores que contribuyeron al problema. Pero solo debemos incluir información relevante y no datos detallados que puedan sobrecargar a la audiencia.

Analizamos los datos y descubrimos que las ventas de "chips" aumentaron un 20%. Sin embargo, las ventas de artículos dulces han disminuido en general en un 30%.



También podemos proporcionar más pruebas y datos de apoyo, siempre y cuando ayudemos a explicar en un nivel más profundo la causa del problema.

Entonces, después de echar un vistazo más profundo, vimos que el chocolate más famoso de la compañia ha visto disminuir sus ventas en un 50%.


Toda la evidencia debería conducir al clímax: el momento en que introducimos el hallazgo central de nuestro análisis, deberia proporcionar información clara sobre lo que podría suceder si no se cambia nada.

Predijimos que si la empresa no toma una decisión, perderá 10 millones de dólares el próximo año.



Después de que se revela el hallazgo principal, deberíamos terminar de explorar podibles soluciones y oportunidades, recomendando un curso de acción a tomar.

Necesitamos ser proactivos y guir a la audiencia a través de la comprensión qué hacer con nuestros resultados si queremos impactar en el proceso de toma de decisiones. 

Entonces, de acuerdo a nuestras predicciones, cambiar la marca de chocolate y ofrecer descuentos especiales puede lograr una ganancia del 20%.

Con estos pasos, creamos una narrativa que captará el interés de nuestra audiencia.



## Construyendo narrativa

Al ejecutar un análisis, hay varias cosas que puede observar, que a su vez impulsarán su narrativa.

**Cambios a lo largo del tiempo:** Podemos explorar cómo cambia una característica con el tiempo o el ciclo de tiempo patrones, por ejemplo, cómo las ventas de chocolate con más bajas en verano pero más altas en invierno.

**Correlación:** Podríamos centrarnos en cómo dos cosas se relacionan entre sí, por ejemplo el precio del chocolate y la calificacion del clientes. 

**Comparación:** o mostrar similitudes y diferencias entre el consumo de chocolate entre adultos y niños.

**Custering:** Además, podemos encontrar grupos en los datos, como altos consumidores de chocolate y altos consumidores de café vs. bajos consumidores de chocolares y bajos consumidores de café.


A medida que elaboramos nuestra narrativa, debemos asegurarnos de que estos puntos de datos hagan una historia relevante y reconocible.

## Variaciones de datos

A Veces queremos comparar una o varias variables específicas a lo largo del tiempo. La diferencia se puede expresar con un número absoluto.

**Absoluto:**

   - Diferencia de ventas entre 2018 y 2017
   
    (3000 unidades más vendidas) 
    
   - El cambio absoluto y el cambio relativo dependen de la cantidad.
   
   
**Relativa:**

   - Variación porcentual 2018 respecto a 2017

    una tasa de crecimiento (10% más de unidades vendidas).

   - Números pequeños más significativos que la realidad.


Si nos enfocamos en una variable, un número absoluto está bien, pero cuando miramos varias variables, la tasa de crecimiento tiende a ser más informativa ya que permite compararlas en diferentes escalas (productos vendios en volúmentes de miles y millones, por ejemplo).

Por ejemplo, para el cambio absoluto, nos importa la diferencia entre las venta de 2018 y 2017.

## Agregados


- Valor representativo:
    
    - Totales / recuentos
    - Medias
    - Mediana

- La media puede ser engañosa (Outlier)

- Distribucion de los datos

A veces necesitamos resumir datos numéricos en un agregado: un número que da una idea de un conjunto o valor representativo.

Puede ser un simple total o recuento, como las ventas totales o cuánto durará una campaña.

Otro agregado común es la media. El número promedio de chocolares o chips vendidados por año, como vemos en el gráfico.

<img src=".\img_2\15.png" width="300px" height="300px">

O el precio medio de cada producto.

<img src=".\img_2\16.png" width="300px" height="300px">

La media puede ser engañosa, especialmente si hay valores atípicos (valores extremos, los datos no se distribuyen normalmente).
En estos casos, la mediana es una métrica más robusta.

Por ejemplo, el salario anual primedio en 2019 en EE.UU. fue de 51916,27 USD, y salario media anual fue de 34248,45 USD. Usando la media, pensariamos que la tendencia común es un salario de 52K, cuando en realidad la mitad de la población está por debajo de $34K.


## p-value

- Qué es p-value?

    - Valor inferior a 0,05: significación estadística
    - Valor próximo a 0,05: indicador débil

- ¿Qué no es?

    - No es una prueba de evidencia:
        -Rechazar nuestra hipótesis, pero no que sea falsa
    - Considerar métricas alternativas o complementarias
    
    
Al comunicar nuestros resultados, a menudo necesitamos proporcionar pruebas de que son estadísticamente significativos (p, que no pueden debeser a la aleatoriedad).

Un valor de p inferior a 0.05 se considera un indicador de importancia por convención. Cuanto menor sea por debajo  de 0.5 más fuerte es el indicador.

Sin embargo, no es una prueba de prueba: sólo rechaza o acepta una hipótesis sin decir nada acerca de la verdad de la misma.

Debido a la frecuencia con la que las métricas del valor p se malinterpretan o son confusas (la audiencia), considere métricas alternativas o agregue algunas más en apoyo.

El líder técnico estará interesado en esto, pero también en más detalles sobre el análisis en sí, como la importancia detallada de la característica.

<img src=".\img_2\21.png" width="350px" height="300px">



## Más buenas prácticas

- Princio de Pareto:
    - Datos agregados menos relevantes
    - Incluye chocolate, patatas fritas y otros productos (agregados)

El principio de Pareto establece que la mayoria de los productos provienen de una minoría de insumos. Así que fuera de los datos que contribuyen a nuestra historia, habrá algunos datos que son menos relevantes pero que queremos incluir. Deberíamos agregarlo para reducir el ruido, como presentado las ventas para chocolates, chip y otros productos en lugar de cada producto por separado.

- Imágenes atractivas y cercanas

Nuestras imágenes pueden volverse accesibles y atractivas si considerando los antecedentes de nuestra audiencia y simplificamos las imágenes al nivel de conocimiento de la audiencia.

- Cuántos / con qué rapidez

Básicamente, le imoprta cuánto conocimiento obtiene su audiencia de su visualización y que tan rápido lo obtienen.

Una trama compleja da muchas ideas, pero toma tiempo para comprender; una trama simple da pocas ideas pero es rápida de entender. 

- Menos es más

En general, menos es más. En lugar de mostrar muchas imágenes detalladas, deberíamos simplemente concéntrarnos en las más simples y relevantes que respaldan nuestro mensaje.


## Método McCandless


1. Presentar la visualización por su nombre

   - Titular del gráfico 
   - Claro y obvio
   - Técnica y vs x

<img src=".\img_2\22.png" width="300px" height="300px">

Hay algunos pasos que podemos seguir para incluir y presentar la visualización de manera efectiva a nuestra audiencia.

Fueron establecido por David McCandless, un periodista de datos británico.

Primero, debemos darle a nuestro gráfico un título, que luego usaremos para Introducir la visualización y centrar la atención de la audiencia en ella. El titular debe ser breve, claro y obvio: apoya nuestra historia y aclara la visualización.

En caso de duda, podemos utilizar la técnica del eje Y frente al eje X, por ejemplo: "ventas de chocolares por mes".


2. Anticiparse a las preguntas del público

   - Centrarse en la historia y no en el gráfico de descodificación.

Luego, debemos anticiparnos a las preguntas comunes de nuestra audiencia, como "¿De dónde provienen estos datos?" o "¿Por qué te enfocaste específicamente en esta característica?".

Con estas preguntas respondidas incluso antes de que se hagan, la atención se centra en nuestro gráfico y nuestra historia.


3. Perspectivas del Estado

Después de eso, debemos aclarar qué información muestra el elemento visual. 

Debemos explicar lo que están viendo y no asumir que lo preguntarán más tarde o que lo entenderán por sí mismos.

Por ejemplo, las ventas de chocolate han disminuido con el tiempo.

<img src=".\img_2\23.png" width="300px" height="300px">


4. Ayudar al público a relacionarse

   - Importancia.
   - Acciones

Por último, necesitamos ayudar a la audiencia a relacionarse con el gráfico y sus ideas. Asegúrese de que entiendan por qué la información es importante, cómo apoyará más puntos de vista en la presentación, o cómo se puede actuar sobre este punto de vista.

Por ejemplo, las ventas de chocolate son relevantes para el departamente de finanzas, ya que tiene un impacto en los ingresos o en el departamento de marketing, ya que la publicidad podría ayudar.

# Elegir el formato adecuado


## El camino de la narración de datos

Esta es la pieza final de nuestra estrategia de comunicación.

Hemos diseñado nuestra historia, hemos definido un enforque, seleccionó los datos correctos, y hemos elegido una visualización impactante.

Ahora, es el momento de presentarlo a nuestro público objetivo.

<img src=".\img_2\24.png" width="400px" height="300px">


## ¿Qué formato es más eficaz?

Un buen formato de comunicación muestra información clave de nuestro proyecto de una manera que sea atractivo, y fácil de entender. 

Hay dos formatos principales que discutirems aqui: 

1. Informes escritos
2. Presentaciones orales

La mayoria de los proyectos de ciencia de datos requerirán un informe escrito y una presentación oral, pero al final depender de la situación y el proyecto en cuestión.


## Estrategia de presentación

Hay varias cosas a tener en cuenta al compartir los hallazgos:

Nuestra audiencia, contenido a incluir, requisitos especiales a tener en cuenta, y qué canal usar. Toos estos elementos nos ayudan a definir el mejor formato para comunicar nuestros resultados.

<img src=".\img_2\25.png" width="400px" height="300px">



## Partes interesadas

+ Quién?

Lo primero que hay que tener en cuenta es a quién nos estamos presentando. Esto ayuda a averiguar por qué necesitan saber acerca de los hallazgos: ¿Por la rendición de cuentas?,¿Para entender la metodologia?. 

+ Cómo?

¿Cómo van a utilizar nuestros hallazgos?, ¿Tomar una decisión?, ¿Para empezar un nuevo proyecto?.

+ Qué?

¿Qué información necesitan?, ¿Nuestros resultados?, ¿El impacto que tienen nuestros hallazgos?


## Contenido

La respuesta a estas preguntas nos ayudará a decidir qué contenido incluir. 

¿Deberiamos centrarnos en los resultados, conclusiones, incluir recomendaciones, o simplemente explicar los métodos en detalle?



## Requerimientos

El siguiente paso es identificar si algún actor tiene requisitos especiales.

* Tiempo

¿Tiene tiempo suficiente para leer un informe detallado o es mejor una reunión breve?

* Autoridad

¿Reportan a otra persona y necesitan un documento para respaldar un reclamo?

* Zona horaria

¿Están en una zona horaria diferente, por lo que se prefiere una comunicación escrita?



## Consumo

También debemos pensar en cómo se consumirá nuestra presentación.

* Formato

Podríamos escribir un documento como un documento de Word, un cuaderno Jupyter, una publicación de blog o un artículo.

* Mecanismo de entrega

También podríamos construir una plataforma de diapositivas. Entonces, ¿cómo se entregará? ¿Presentaremos directamente a las partes interesadas, pudiendo responder a los comentarios o preguntas, o lo compartiremos por correo electrónico o Slack y escribiremos preguntas más tarde? 

* Audiencia

Finalmente, ¿qué tan grande sería la audiencia?. No presentarías las cosas de la misma manera en una sala de conferencias de seis personas o en un salón de baile de doscientas personas.



## Comunicación Oral

Seleccionar el formato correcto depende no solo de la audiencia, sino también en las fortalezas y debilidades de los diferentes formatos.

* Ventajas:

La comunicación oral permite construir una relación con la audiencia, y para comentarios o acciones inmediatas.

Entrega un mensaje rico porque el lenguaje corporal y la voz agregan significado. 

* Desventajas:

Sin embargo, a menos que proporciones una grabación, el mensaje no se puede revisar porque no hay registro permanente de comunicación, y no es adecuado para mensajes extensos, ya que más larga es la presentación, más posibilidades hay de que la audiencia pierda el foco en algún momento.


## Comunicación Escrita

* Ventajas:

El formato escrito proporciona registro de comunicación para que el mensaje pueda analizarse a largo plazo.

Es fácil de compartir con grandes audiencias, y es menos propenso a las reacciones emocionales.

También es adecuado para compartir el código con cualquier parte interesada técnica para su revisión o replicación.


* Desventajas:

Sin embargo, no ver la reacción de la audiencia hace que sea más difícil adaptarse, ya que la retroalimentación no es inmediata sino que vendrá más tarde en forma de comentarios.



## Formato adecuado

**Presentación Oral:**

Veamos un ejemplo.

* ¿A quién? **CEO**
* ¿Por qué? **Actualización mensual**
* ¿Cómo? **Rendición de cuentas**
* Contenido: **Conclusiones**
* Requisito: **Tiempo**
* Canal: **Reunión**

Estamos presentado nuestros hallazgos a el director general de la empresa communicatb. está interesada en nuestras conclusiones y quiere respuestas rápidas a sus preguntas.

Está bastante ocupada, pero sucede que visita nuestra oficina al final de la semana. Esta es una ocasión perfecta para organizar una reunión rápida y presentar nuestros hallazgos mediante una presantación oral.


**Reporte**

* ¿A quién? **CEO**
* ¿Por qué? **Próximas decisiones importantes**
* ¿Cómo? **Informe a los inversores**
* Contenido: **Recomendaciones**
* Canal: **Email**

Después de la reunión, el director general nos dice que decisiones importantes se basarán en nuestras conclusiones y reportado a los inversionistas. 

Ahora es el momento de estructurarlos en un informe escrito, teniendo en cuenta las preguntas y los comentarios que recibimos de ella.
