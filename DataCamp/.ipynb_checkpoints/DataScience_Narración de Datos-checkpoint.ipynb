## ¿Por qué son necesarias las historias?

Las personas que trabajan con datos pasan mucho tiempo recopilando y procesando datos, y modelando y evaluando resultados.

Pero, ¿qué pasa con la comunicación de ideas?

Incluso los mejores resultados podrían no tener un impacto se las ideas no se presentan de manera adecuada.

Uno de los principales desafíos es convencer a las parte interesadas adversas al cambio de que se deben hacer algunas cosas diferentemente, especialemte si estas partes interesadas no son técnicas y no se sienten cómodos con las matemáticas o las estadísticas.

## ¿Qué es la narración de datos?

Por eso, contar historias con datos es primordial. Entonces, que es la narración de datos?

Es un mecanismo poderoso para compartir información respaldada por una narrativa convincente y visualizaciones eficientes.

**¿Por qué es tan poderoso?**

Las anécdotas impulsan la imaginación de las personas y las historias son más memorables que las métricas y agregan valor al dar contexto a los datos. Gracias a eso, la audiencia puede entender ideas. Así se capta su atención.

Esos conocimientos facilitan el proceso de toma de decisiones y, a su vez, impulsar acciones y cambios.



## Narración de Datos  (Data Storytelling)

Comencemos con la construcción de una historia de datos efectiva.

Dos conceptos nos ayudan a construir su núcleo.

El primer concepto, la historia de tres minutos, nos hace pensar: si solo teníamos tres minutos para contar una historia, ¿en qué nos enfocaríamos?

El segundo concepto, la gran idea, nos empuja a expresar el punto de vista único de nuestra historia en una oración.

Estos conceptos nos ayudan a articular nuestra historia de manera clara y concisa.


Hay tres elementos centrales para cualquier historia de datos:

Estos elementos hacen la historia 
- **Perspicaz** --> ya que deriva aprendizajes claros de nuestro análisis. 
- **Explicativos** --> porque ayudamos a nuestra audiencia a comprender sus ideas.
- **conciso** --> dando sólo hechos concretos y especificos.

## Narración convincente

Una vez qye hayamos encontrado información relevante de nuestros datos, tenemos que encontrar una manera significativa de presentarlos a nuestro público objetivo, usando una narrativa que incluye solo puntos clave  y nos ayuda a impulsar el cambio.

En otras palabras, necesitamos una narrativa convincente o una descripción de eventos conectados que organiza la información para involucrar a la audiencia y hacer que se interesen por los resultados o la información compartida. ## Estructura Narrativa


Para comenzar nuestra historia de datos, debemos mencionar los antecedentes:

   - Detalles sobre qué problemas motivó el análisi.
   - Lo que cambió en la situación previa que ameritaba un análisis.
   - Y en quién se centra el análisis.
       - ¿Estamos analizando los datos de los clientes?, De los empleados?. o de algo más?
       
       
En nuestro caso, mencionamos que frente a las expectativas de la gerencia las utilidades totales de la compañia han disminuido en los últimos tres trimestres.


Después de que se introdujo el problema, debe proporcionar evidencia de los factores que contribuyeron al problema. Pero solo debemos incluir información relevante y no datos detallados que puedan sobrecargar a la audiencia.

Analizamos los datos y descubrimos que las ventas de "chips" aumentaron un 20%. Sin embargo, las ventas de artículos dulces han disminuido en general en un 30%.



También podemos proporcionar más pruebas y datos de apoyo, siempre y cuando ayudemos a explicar en un nivel más profundo la causa del problema.

Entonces, después de echar un vistazo más profundo, vimos que el chocolate más famoso de la compañia ha visto disminuir sus ventas en un 50%.


Toda la evidencia debería conducir al clímax: el momento en que introducimos el hallazgo central de nuestro análisis, deberia proporcionar información clara sobre lo que podría suceder si no se cambia nada.

Predijimos que si la empresa no toma una decisión, perderá 10 millones de dólares el próximo año.



Después de que se revela el hallazgo principal, deberíamos terminar de explorar podibles soluciones y oportunidades, recomendando un curso de acción a tomar.

Necesitamos ser proactivos y guir a la audiencia a través de la comprensión qué hacer con nuestros resultados si queremos impactar en el proceso de toma de decisiones. 

Entonces, de acuerdo a nuestras predicciones, cambiar la marca de chocolate y ofrecer descuentos especiales puede lograr una ganancia del 20%.

## Construyendo narrativa

Al ejecutar un análisis, hay varias cosas que puede observar, que a su vez impulsarán su narrativa.

**Cambios a lo largo del tiempo:** Podemos explorar cómo cambia una característica con el tiempo o el ciclo de tiempo patrones, por ejemplo, cómo las ventas de chocolate con más bajas en verano pero más altas en invierno.

**Correlación:** Podríamos centrarnos en cómo dos cosas se relacionan entre sí, por ejemplo el precio del chocolate y la calificacion del clientes. 

**Comparación:** o mostrar similitudes y diferencias entre el consumo de chocolate entre adultos y niños.

**Custering:** Además, podemos encontrar grupos en los datos, como altos consumidores de chocolate y altos consumidores de café vs. bajos consumidores de chocolares y bajos consumidores de café.


A medida que elaboramos nuestra narrativa, debemos asegurarnos de que estos puntos de datos hagan una historia relevante y reconocible.

A Veces queremos comparar una o varias variables específicas a lo largo del tiempo. La diferencia se puede expresar con un número absoluto.

**Relativa:**

## Agregados


- Valor representativo:
    
    - Totales / recuentos
    - Medias
    - Mediana

- La media puede ser engañosa (Outlier)

- Distribucion de los datos

A veces necesitamos resumir datos numéricos en un agregado: un número que da una idea de un conjunto o valor representativo.

Puede ser un simple total o recuento, como las ventas totales o cuánto durará una campaña.

Otro agregado común es la media. El número promedio de chocolares o chips vendidados por año, como vemos en el gráfico.

<img src=".\img_2\15.png" width="300px" height="300px">

O el precio medio de cada producto.

<img src=".\img_2\16.png" width="300px" height="300px">

La media puede ser engañosa, especialmente si hay valores atípicos (valores extremos, los datos no se distribuyen normalmente).
En estos casos, la mediana es una métrica más robusta.

Por ejemplo, el salario anual primedio en 2019 en EE.UU. fue de 51916,27 USD, y salario media anual fue de 34248,45 USD. Usando la media, pensariamos que la tendencia común es un salario de 52K, cuando en realidad la mitad de la población está por debajo de $34K.


## p-value

- Qué es p-value?

    - Valor inferior a 0,05: significación estadística
    - Valor próximo a 0,05: indicador débil

- ¿Qué no es?

    - No es una prueba de evidencia:
        -Rechazar nuestra hipótesis, pero no que sea falsa
    - Considerar métricas alternativas o complementarias
    
    
Al comunicar nuestros resultados, a menudo necesitamos proporcionar pruebas de que son estadísticamente significativos (p, que no pueden debeser a la aleatoriedad).

Un valor de p inferior a 0.05 se considera un indicador de importancia por convención. Cuanto menor sea por debajo  de 0.5 más fuerte es el indicador.

Sin embargo, no es una prueba de prueba: sólo rechaza o acepta una hipótesis sin decir nada acerca de la verdad de la misma.

Debido a la frecuencia con la que las métricas del valor p se malinterpretan o son confusas (la audiencia), considere métricas alternativas o agregue algunas más en apoyo.

El líder técnico estará interesado en esto, pero también en más detalles sobre el análisis en sí, como la importancia detallada de la característica.

## Más buenas prácticas

- Princio de Pareto:
    - Datos agregados menos relevantes
    - Incluye chocolate, patatas fritas y otros productos (agregados)

El principio de Pareto establece que la mayoria de los productos provienen de una minoría de insumos. Así que fuera de los datos que contribuyen a nuestra historia, habrá algunos datos que son menos relevantes pero que queremos incluir. Deberíamos agregarlo para reducir el ruido, como presentado las ventas para chocolates, chip y otros productos en lugar de cada producto por separado.

- Imágenes atractivas y cercanas

Nuestras imágenes pueden volverse accesibles y atractivas si considerando los antecedentes de nuestra audiencia y simplificamos las imágenes al nivel de conocimiento de la audiencia.

- Cuántos / con qué rapidez

Básicamente, le imoprta cuánto conocimiento obtiene su audiencia de su visualización y que tan rápido lo obtienen.

Una trama compleja da muchas ideas, pero toma tiempo para comprender; una trama simple da pocas ideas pero es rápida de entender. 

- Menos es más

En general, menos es más. En lugar de mostrar muchas imágenes detalladas, deberíamos simplemente concéntrarnos en las más simples y relevantes que respaldan nuestro mensaje.


## Método McCandless


1. Presentar la visualización por su nombre

   - Titular del gráfico 
   - Claro y obvio
   - Técnica y vs x

<img src=".\img_2\22.png" width="300px" height="300px">

Hay algunos pasos que podemos seguir para incluir y presentar la visualización de manera efectiva a nuestra audiencia.

Fueron establecido por David McCandless, un periodista de datos británico.

Primero, debemos darle a nuestro gráfico un título, que luego usaremos para Introducir la visualización y centrar la atención de la audiencia en ella. El titular debe ser breve, claro y obvio: apoya nuestra historia y aclara la visualización.

En caso de duda, podemos utilizar la técnica del eje Y frente al eje X, por ejemplo: "ventas de chocolares por mes".


2. Anticiparse a las preguntas del público

   - Centrarse en la historia y no en el gráfico de descodificación.

Luego, debemos anticiparnos a las preguntas comunes de nuestra audiencia, como "¿De dónde provienen estos datos?" o "¿Por qué te enfocaste específicamente en esta característica?".

Con estas preguntas respondidas incluso antes de que se hagan, la atención se centra en nuestro gráfico y nuestra historia.


3. Perspectivas del Estado

Después de eso, debemos aclarar qué información muestra el elemento visual. 

Debemos explicar lo que están viendo y no asumir que lo preguntarán más tarde o que lo entenderán por sí mismos.

Por ejemplo, las ventas de chocolate han disminuido con el tiempo.

<img src=".\img_2\23.png" width="300px" height="300px">


4. Ayudar al público a relacionarse

   - Importancia.
   - Acciones

Por último, necesitamos ayudar a la audiencia a relacionarse con el gráfico y sus ideas. Asegúrese de que entiendan por qué la información es importante, cómo apoyará más puntos de vista en la presentación, o cómo se puede actuar sobre este punto de vista.

Por ejemplo, las ventas de chocolate son relevantes para el departamente de finanzas, ya que tiene un impacto en los ingresos o en el departamento de marketing, ya que la publicidad podría ayudar.

## ¿Qué formato es más eficaz?

Un buen formato de comunicación muestra información clave de nuestro proyecto de una manera que sea atractivo, y fácil de entender. 

Hay dos formatos principales que discutirems aqui: 

1. Informes escritos
2. Presentaciones orales

La mayoria de los proyectos de ciencia de datos requerirán un informe escrito y una presentación oral, pero al final depender de la situación y el proyecto en cuestión.


## Estrategia de presentación

Hay varias cosas a tener en cuenta al compartir los hallazgos:

Nuestra audiencia, contenido a incluir, requisitos especiales a tener en cuenta, y qué canal usar. Toos estos elementos nos ayudan a definir el mejor formato para comunicar nuestros resultados.

¿Cómo van a utilizar nuestros hallazgos?, ¿Tomar una decisión?, ¿Para empezar un nuevo proyecto?.

La respuesta a estas preguntas nos ayudará a decidir qué contenido incluir.

* Zona horaria

## Consumo

También debemos pensar en cómo se consumirá nuestra presentación.

* Formato

* Mecanismo de entrega

También podríamos construir una plataforma de diapositivas. Entonces, ¿cómo se entregará? ¿Presentaremos directamente a las partes interesadas, pudiendo responder a los comentarios o preguntas, o lo compartiremos por correo electrónico o Slack y escribiremos preguntas más tarde? 

* Audiencia

Finalmente, ¿qué tan grande sería la audiencia?. No presentarías las cosas de la misma manera en una sala de conferencias de seis personas o en un salón de baile de doscientas personas.



## Comunicación Oral

Seleccionar el formato correcto depende no solo de la audiencia, sino también en las fortalezas y debilidades de los diferentes formatos.

* Ventajas:

La comunicación oral permite construir una relación con la audiencia, y para comentarios o acciones inmediatas.

Entrega un mensaje rico porque el lenguaje corporal y la voz agregan significado. 

* Desventajas:

Sin embargo, a menos que proporciones una grabación, el mensaje no se puede revisar porque no hay registro permanente de comunicación, y no es adecuado para mensajes extensos, ya que más larga es la presentación, más posibilidades hay de que la audiencia pierda el foco en algún momento.


## Comunicación Escrita

* Ventajas:

El formato escrito proporciona registro de comunicación para que el mensaje pueda analizarse a largo plazo.

Es fácil de compartir con grandes audiencias, y es menos propenso a las reacciones emocionales.

También es adecuado para compartir el código con cualquier parte interesada técnica para su revisión o replicación.


* Desventajas:

Sin embargo, no ver la reacción de la audiencia hace que sea más difícil adaptarse, ya que la retroalimentación no es inmediata sino que vendrá más tarde en forma de comentarios.

# Tipos de Reportes

## ¿Cómo estructurar un informe escrito?

- Tipos de reportes
- Reproducibilidad
- Redactar informes precisos y claros.


## Informe Escrito

Imagina que en communicatb, necesitamos escribir un informe explicando nuestro proyecto centrado en análisis de sentimiento e las reseñas de productos.

Nuestro informe debe comunicar nuestros hallazgos con respecto a las calificaciones negativas debido a envíos retradados, y la precisión de la predicción que mostro nuestro modelo. Tenemos que presentar nuestro trabajo de acuerdo a los estándares de la industria, entonces nuestros resultados son validados y nuestras recomendaciones impulsan el cambio esperado.


## Tipos de Reportes

Hay diferentes tipos de informes:

- Informativos: Proporcionan información fáctica, suelens er cortos, no tienen una estructuraa estricta,  y su finalidad es informar sobre hechos sin añadir ningún análisis.

- Analíticos: Proporcionan análisis y demostrar relaciones o recomendaciones. Puede variar en tamaño, pero tiene una estructura estricta,  y su porpósito es impulsar la decisión basada en datos.


### Informes Finales

**Elementos:**

- Incluye análisis de datos detallados, hallazgos y resultados, así como imágenes.

**Formato:**

- Suelen ser largos,

**Audiencia:**

- porque están destinados a audiencias que necesitan detalles técnicos.


### Informes Resumidos

**Elementos:**

- Incluye conclusiones y recomendaciónes clave, así como imágenes.

**Formato:**

- Suelen ser breves, de menos de cinco páginas. Debido a que son un resumen de un informe final, pueden incluir enlaces al documento principal. 

**Audiencia:**

- Están destinados a los tomadores de decisiones que no necesitan detalles técnicos.


## Estructura del informe

Como mencionamos, los informes informativos no siguen una estructura estricta. Pero hay una estructura sencilla que debemos seguir para los informes analíticos, finales y resumidos.


* **Introducción**

    - Propósito
    
    - Información Contextual
    
    
    - Preguntas de análisis
    
La primera sección es la intruducción. A continuación, resumimos el propósito del informe. En nuestro ejemplo, el propósito del informe es mostrar el análisis de las reseñas de productos, y los resultados de las predicciones de calificación. 

Después de eso debemos incluir información contextual sobre por qué realizamos el análisis que se muestra. Lo que motivó nuestro informe de ejemplo es un aumento en las críticas negativas.

Finalmente, debemos resumir nuestras preguntas de análisis. Preguntamos qué factores están impulsando la mala experiencia del usuario.


* **Cuerpo**

    - Datos:
        - Descripcion y tablas
        
    - Metodos:
        - NLP y Bosque Aleatorio
    
    - Análisis:
        - Visuales
            - Graficos con las palabras más comunes
    - Resultados:
        - Descripción y Visuales
        
        
Luego, creamos las secciones para el cuerpo del informe. En la sección de datos, escribimos una descripción de los datos más relevantes utilizados. Podemos incluir tablas.

En la sección Método, describimos los métodos utilizados para recopilar y analizar los datos y construir el modelo. En nuestro caso, explicamos que usamos procesamiento de lenguaje natural y un modelo de bosque aleatorio.

En la sección Análisis incluimos los datos seleccionados para el análisis y modelar usando imágenes, por ejemplo, un gráfico con las palabras más comunes.

En la sección de resultados, describimos y explicamos los resultados de nuestro análisis. Tambien podriamos incluir imágenes para evaluarlos. Por ejemplo, afirmar que el 30% de las calificaciones negativas se asociaron con los términos "retrado" y  "envio".


* **Conclusiónes**

    - Replantear Preguntas
    - Resumir los resultados importantes
    - Agregar Recomendaciónes
    
Finalmente creamos una sección de conclusiones. Aquí, replanteamos las preguntas de análisis, así como resumir los resultados más importantes del análisis.

Esta parte es la ubicación adecuada para agregar nuestras recomendaciones para los próximos pasos.

---


* **Contexto empresarial**

La estructura que acabamos de ver es adecuada para informes de datos en revistas. En un contexto empresarial, su audiencia es diferente, por lo que debe adaptarse. Un método eficientees el enfoque 1-3-25:

* **1-3-25**
    - 1 página de resumen
    - 3 páginas máximo de resumen ejecutivo
    - 25 páginas como máximo de detalles.
    
El resumen ejecutivo es suficiente para que la gente entienda sus conclusiones y decidir si vale la pena leer el informe en su totalidad.


## Audiencia

Una vez más, debemos tener en cuenta a nuestra audiencia. Cada parte interesada está interesada de diferentes partes de nuestro informe. Así que debemos adaptarlo en consecuencia. Por ejemplo:

* Personas con poco tiempo:
    - Introducción
    - Conclusión
    - Escanea el cuerpo

Las personas con poco tiempo pero interesado en el tema leerá la introducción y la conclusión y luego escaneará el cuerpo en busca de puntos particulares.


* Cliente o colaborador interno:
* Equipo ejecutivo:
    - Leerá introducción y conclusiones
    - Recomendaciónes
    
El equipo ejecutivo leerá rápidamente la introducción y la conclusión para encontrar pistas relacionadas con su principal interés: nuestras recomendaciones.


* Parte interesada técnica:
    - Cuerpo
    
Finalmente, una parte interesada técnica está principalmente interesada en el cuerpo del informe. quiere comprender y validar nuestro métodos y análisis técnicos.

# Reproducibilidad y referencias

## Informe escrito

Una parte fundamental de la comunicación de nuestros hallazgos, es asegurarse de que un informe sea claro y reproducible.

## Ejemplo de reproducibilidad

Por ejemplo, digamos que tenemos un amigo que siempre cocina este increíble paste de chocolate. Llegamos a su casa y seguimos sus pasos exactos con su equipo: eso es replicabilidad. Confiamos en que realmente puede hornear este delicioso pastel.

En un proyecto de datos, si vuelve a ejecutar su Jupyter Notebook hoy y mañana, mis resultados deberían ser idénticos (siempre que el conjunto de datos no haya cambiado).

---

El horneado, la reproducibilidad sería la capacidad de volver a cocinar ese pastel yo mismo, siguiendo la receta pero usando mis propios utensilios e ingredientes.

La reproducibilidad se refiere a la capacidad de obtener resultados similares a los de otro equipo pero utilizando datos y códigos diferentes. 

En un proyecto de datos, sería la capacidad de obtener resultados similares utilizando el mismo enfoque general, pero un entorno diferente (los datos procesan de la misma manera, pero utilizando un idioma diferente en un sistema operativo diferente).


## Virtudes de reproducibilidad y replicabilidad

Este enfoque científico es crucial porque evita la duplicación de esfuerzos y permite otros cientificos de datos a contruir sobre el trabajo preexistente, permitiendoles construir sobre el trabajo previo y enforcarse en nuevos desafíos. Por último, pero no menos importante, permite la revisión por pares.

Aunque tendemos a centrarnos en ejemplo de codificación en este curso, siempre debe asegurarse de que código sea fácilmente reproducible y replicable, ya sea que usemos hojas de cálculo de Excel, visualizaciones de Tableay o Jupyter Notebooks. Lo que importa es hacer un seguimiento de cómo se produjeron los resultados.



## Buenas prácticas


1. **llevar la cuenta de cómo se han producido los resultados.**

2. **Evitar la manipulación manual de los datos.**

3. **Documentar la aleatoriedad.**

4. **Interpretabilidad.**

5. **Citar bibliografía correctamente.**


Deberíamos documentar todos los scripts que usamos para obtener nuestros resultados, usando por ejemplo, comentarios en nuestro código, y enumere todos los paquetes utilizados en nuestro entorno. Una tecnica útil es utilizar un sistema de control de versiones (como Git), que rastrea todos los cambios y versiones de nuestros scripts. 

Debemos evitar hacer cualquier manipulación manual de los datos. Nunca debemos cambiar el conjunto de datos directamente en un editor. Si es posible, debemos guarar todas las versiones de nuestro conjunto de datos. Además, debemos guardar los datos sin procesar junto con un script con pasos intermedios. Puede ayudarnos a contar la historia de nuestra transformación de datos y crear una narrativa a su alrededor.

Una visión tan clara asegura que sepamos en el punto lo que esta sucediendo con los datos, y por lo tanto pueden adaptarse y resolver problemas.

Tomemos un ejemplo de imputación de datos. Hubo un error en la canalización de datos y se le dice que puede usar los valores medios para imputar valores faltantes para un producto específico. Continúe y presione las ediciones manualmente, luego cierre su editor. Inmediatamente después, su colega le informa que, en realidad, esos productos no estaban disponibles para la venta en esas fechas, por lo que los valores faltantes deberían ser cero. Los datos ya están sobreescritos, por lo que será dificil saber qúe valores se cambiaron en primer lugar. si hubiéramos versionado los cambios, sería mucho más sencillo.


Usualmente usamos algoritmos de aprendizaje automático o canalizaciones para crear nuestro flujo de trabajo, algunos de ellos podrían involucrar técnicas de aleatoriedad. Por lo general, podemos establecer una semilla aleatoria e introducir reproducilidad en los resultados de nuestro modelo. 

La semilla aleatoria controla las variables de confusión: nos permite asegurar que un cambio se debió al modelo y no solo a la aleatoriedad. 

Interpretabilidad es el grado en que un ser humano puede entender la causa de una decisión o puede predecir consistentemente el resultado de un modelo. Contar una historia con una narrativa convincente ayuda a nuestros grupos de interés a comprender e interpretar nuestros hallazgos. Debido a que son capaces de entederlos, nuestras conclusiones pueden reproducirse. 

Por último, siempre debemos citar correctamente el trabajo de otras personas utilizado en nuestro análisis.


## Referencias

Las citas son la información básica requerida para identificar y ubicar una publicación.

Hay diferentes estilos, pero tosos tiene la misma lógica subyacente, con ligeras variaciones.

El estilo más común es APA, que utiliza citas en el texto, poniendo primero el nombre del autor y luego la fecha de publicación.

Hay una serie de herramientas de gestión de referencias, eso puede aliviar la carga de rastrear todas las citas. Nos ayudan a cambiar estilos automáticamente, y busque fuentes en línea para obtener referencias. `EndNote`, `Mendeley` y `RefWorks` son algunos ejemplo, pero hay muchos más.

---


En un contexto empresarial, las reglas de estas ediciones son mucho más relajadas. La mayoria de la gente simplemente incluirá un hi´pervinculo a la fuente. Lo que importa el final es que la información sea fácil de recuperar, si el lector quiere consultar el material de las fuentes.

# Escribir informes precisos y claros.


## Informe escrito

Para tener un impacto a nivel de decisión, otras partes interesadas deben validar nuestro trabajo. Por lo tanto, nuestro informe no solo debe ser reproducible, sino también claro e inteligible.


## Redactar informes precisos y claros

En ciencia de datos, la escritura debe ser conciso y preciso, para evitar engaños y confusiones. Partes interesadas deben entender nuestro mensaje.

En nuestros informes, no hay lugar para oraciones que no agreguen información o introduzcan posibles malentendidos.

Exploremos alguna opciones para ayudarnos con esta tarea.


## Frases vacías

Las frases vacías básicamente no agregan significado o información. 

Frases como: 

- Es interesante notar que ...
- El hecho de que ...
- Cabe señalar que ...
- Es bien sabido que ...
- Es obvio que ...


Las frases vacías distraen y deben evitarse: sea directo y directo.

Aquí, ambas oraciones transmiten el mismo mensaje, pero la primera va directo al grano, mientras que el segundo es más difícil de seguir.

> Las valoraciones negativas se asocian a las palabras "retraso" y "envío".


> Otro dato importante es que las valoraciones negativas se asocian a las palabras "retraso" y "envío".

Básicamente, si no agrega información, elimínela.


## Sustantivos concretos

La buena redacción técnica es concisa y directa. Así que deberiamos escribir sustantivos concretos, y evitar el uso excesivo de los pronombres "Esto", "Eso" o "Es", ya que pueden no estar claro a qué se refieren exactamente.

Además, hacer un seguimiento de lo que significan los pronombres agregar cierta carga cognitiva y distrae a los lectores de la realidad.

En la oración de ejemplo: 

> Esto demuestra una precisión del 80% a la hora de predecir la pérdida de clientes.

La palabra "esto" debería ser reemplazdo por un sustantivo concreto y que afirme que:

> El modelo muestra una precisión del 80% a la hora de predecir la pérdida de clientes.



## Más pronombres


La voz activa es criticada por poner el énfasis en el autor en lugar de los hechos.

La voz pasiva es criticada por ser demasiado cargada y difícil de leer.

Si se prefiere la voz pasiva en un entorno académico, en un contexto de negocios, su mensaje probablemente llegará mejor usando la voz activa.


## Adjetivos y adverbios redundantes


Cuando se trata de enfatizar un argumento, es fácil terminar usando adjetivos y adverbios redundantes. 

Algunos ejemplos son: 

- La introducción de los nuevo
- Hecho previamente

Para que el informe sea conciso, debemos eliminar estos adjetivos o adverbios redundantes.


## Frases incoherentes


Por último, Dos o más oraciones o cláusulas independientes conectadas o puntuadas incorrectamente. 

Por ejemplo, la oración:

> Existe una correlación entre el retraso en el envío y la valoración del cliente, el retraso en el envío es la causa de la valoración negativa.

Se prolonga. Se compone de dos oraciones completas pero conectadas por una coma. 

Para arreglar estas oraciones, deberíamos hacer dos clausulas (o oraciones) por separado o usar cláusulas dependientes intruduciendo palabras como "porque" o "así".

# Planificación de una presentación oral


## Estrategia de Presentación

<img src=".\img_2\25.png" width="400px" height="300px">

Ahora sabemos qué elementos debemos considerar al elegir el formato apropiado para entregar nuestros resultados. Nos enfocamos en informes escritos, ahora dirijamos nuestra atención a la presentación oral.


## Plan de presentación


* Antes de construir diapositivas: Planifica!

* Estructura de Presentación:
    - Propósito
    - Audiencia
    - Mensaje

Así como es una mala idea sumergirse directamente en la escritura de código para un análisis de datos sin primero haber pensado en lo que quiere hacer, por lo general es una mala idea sumergirse en la escritura de diapositivas sin planificar.

Para ayudar a estructurar la presentación, debemos considerar un propósito, quién es la audiencia, y el mensaje que queremos transmitir.


## Propósito

Volvamos a nuestro análisis de opinión de las reseñas de producto que resumimos en un informe en el Capitulo 3.

Para definir el propósito de la presentación, consideramos el tipo de presentación que estamos dando.

**¿Qué tipo de presentación?**

- **Informativa** (¿Es una presentación informativa?): Si es así, es posible que deseemos informar sobre el número actual de calificaciones negativas y positivas y palabras asociadas con críticas negativas.

- **Instrucción** (¿Estamos tratando de instruir a nuestros colegas o compañeros de trabajo?): Entonces quizás queramos explicar cómo se contruyó el modelo para el análisis de sentimientos.

- **Persuasivo** (¿O estamos tratando de persuadir a algunas partes interesadas?): Nuestro propósito sería entonces convencer a las partes interesadas de que es necesario tomar acciones para revertir el alto número de calificaciones negativas.



## Audiencia


Después de entender el propósito de nuestra presentación, debemos pensar sobre quién es nuestra audiencia.

**¿Quién es el público?**

- **¿Estamos presentando a un colega técnico?**

- **¿A un gerente o equipo ejecutivo?**

- **¿O a un cliente?**


Pero también debemos pensar cuan grande es nuestra audiencia. ¿Estamos presentando en una pequeña reunión, por ejemplo, a 10 miembros del equipo financiero?¿O a es una gran reunión o una conferencia? 

Por ejemplo, ¿una reunión con 100 empleados del departamento de desarrollo de software?. O tal vez estamos dando un taller, como una formación técnica para 30 empleados del cliente de sus respectivos departamentos de TI?



## Mensaje

Nuestra audiencia habrá olvidado el 90% de la presentación una semana después de verla. Si solo recuerdan una cosa de tu presentación, ¿qué quieres que sea?.

Luego puede trabajar hacia atrá a partir de eso para decidir qué debe incluirse en la presentación para que la audiencia llegue allí.

Su declaración de apertura debe enganchar a su audiencia y presentar esto con lo que quiere que la audiencia se vaya. Por ejemplo: las valoraciones negativas ahuyentan a los clientes de nuestro sitio web. 

Luego, decidimos el mensaje central. Es aconsejable usar solo una oración. Podríamos transmitir que el retraso en el envió es la principal causa de las críticas negativas y se necesitan acciones inmediatas para revertir la situación.

Finalmente, necesitamos una declaración de cierre que resuma la presentación y refuerza el mensaje central. Podríamos decir que hay una disminución en las ventas. Las críticas negativas han ido en aumento. El envio retradaso está causando calificaciones negativas y se necesitan accionas para revertir la situación. 




## Estructura


Teniendo todos estos pasos en cuenta, podemos pensar en la estructura. Comenzando con:


* Introducción: Debemos proporcionar algunos antecedentes para captar la atención de nuestra audiencia y darles una idea del contenido de la presentación.

* Despues de eso, podemos sumergirnos en los métodos, el análisis y los resultados del modelo. Quién es nuestra audiencia determina qué tan detallada debe ser esta parte: si nuestra audiencia no es técnica, podría justificar omitir esto por completo.


* Finalmente, concluimos, refiriéndose a la introducción, agregando una declaración de llamado a la acción o recomendaciones para los próximos pasos. Esta estructura ayuda a mantener a nuestra audiencia comprometida, que probablemente esté más atenta al principio y al final de nuestra charla. 


## Describir

Ahora es el momento de delinear los gráficos y visuales que queremos usar, y planifique qué secciones necesitamos y su orden.

Deberíamos tener alrededor de cinco secciones y evitar tener demasiadas más. Puede hacer que sea difícil mantener la atención de nuestra audiencia.

Podríamos tener las siguientes secciones:

1. La razón del análisis.

2. Análisis Exploratorio.

3. Análisis de Sentimiento.

4. Conclusiones.

5. Acciones de Seguimiento.


## Tener en cuenta el tiempo!


Por último, pero no menos importante, el tiempo que tenemos para entregar nuestra presentación afecta la forma en que la planificamos.

# Creación de diapositivas de presentación



## De la planificación a la construcción


* Diapositivas
    - Apoyar la historia.
    - Diapositivas refinadas.
    - Recuento o sincronización del diapositivas = mala métrica.
    - Un mensaje por Diapositiva.


Ahora, debemos diseñar las diapositivas.

Hay algunos sugerencias que podemos seguir. Nuestras diapositivas deben apoyar la historia que estamos transmitiendo. Al construirs las diapositivas, debemos tener en cuenta la capacidad de atención de nuestra audiencia. Es mejor tener muchas diapositivas con poco contenido en cada una, para una presentación más dinámica, en lugar de unas pocas diapositivas con mucho contenido.


Hay una buena razón para contar las diapositivas: la idea es evitar que los presentadores se excedan en el tiempo. Pero no debemos preocuparnos por contar diapositivas o cuánto tiempo dedicamos a cada una.

En lugar deberíamos tener un mensaje por diapositiva. Mantendrá las cosas breve y fluidas. Nuestra audiencia solo puede absorver una cierta cantidad de información a la vez. Tratar de cubrir demasiada información simultáneamente aumentará la carga cognitiva, nubla nuestro mensaje y hace que nuestra audiencia pierda algunos de los puntos clave. 


## Color

* **Transmitir significado:**

El formateo puede enfatizar, o por el contrario nublar, el mensaje que queremos transmitir.

Una técnica de formato es el color. Recuerde, todo en una diapositiva debe transmitir un significado y todo lo demás debe cortarse.

Entonces, en lo que respecta a los colores, usamos la menor cantidad de colores necesarios para transmitir nuestro mensaje. 


* **Legibilidad**

<img src=".\img_2\31.png" width="500px" height="300px">

Además, debemos mantener un buen contraste entre el texto y el fondo, para ayudar a la legibilidad.



* **Inclusivo**
    - Diferenciar colores
    - Ejemplo: verde y rojo

También debemos estar atentos a los auditores potencialmente incapacitados. 

Las personas que tienen deficiencia en la visión del color no pueden distinguir entre dos colores diferentes, por lo que esas personas tendrán dificultades para comprender el texto o el gráfico en las diapositivas si usa verde y rojo juntos.



## Fuentes


Las fuentes también afectan la legibilidad. 

* **Serif vs. Sans-serif**

Los investigadores no se ponen de acuerdo sobre qué tipo de letras es más fácil de leer entre `serif` y `sans-serif`. 


<img src=".\img_2\32.png" width="600px" height="300px">


La regla empírica sería que sans serif es apropiado para texto que leerse en una pantalla, y serif es apropiado para el texto que se imprimirá.

* **Contexto**

La elección de la fuente también es contextual: Comic Sans Ms tiene una mala reputación, pero es una gran fuenta si se presenta a los niños. No tanto en un entorno empresarial.


* **Apoyo**

Al final, el contenido de la diapositiva está ahi para respaldar lo que estamos diciendo. 

* **Tamaño**

Y, por lo general, los textos son difíciles de leer, no por la elección de la fuente, sino porque son textos largos en una fuente pequeña. Mantengamos nuestros textos cortos y grandes en su lugar.


* **Varias fuentes**

También podemos utilizar varia tipografías, cambiar el espaciado de letras y lineas, así como usar negrita, cursiva y varios tamaños de letra. Deberiamos usar la menor cantidad de estos para transmitir nuestro mensaje sin distraernos. 



## Texto de la diapositiva


* **Demasiado texto**
    - La audiencia tiende a leer en lugar de escuchar.

<img src=".\img_2\33.png" width="300px" height="300px">

Incluir demasiado texto puede tener un impacto negativo. La audiencia tiende a leer la diapositiva en lugar de escuchar.

* **Puntos principales**

En su lugar, debemos escribir los puntos principales que respaldan nuestro mensaje. Las diapositivas son solo un apoyo. No están diseñados para ser consumidos sin un presentador: si es necesario, entonces una grabación puede hacer el truco, pero no deberíamos convertir nuestras diapositivas en informes escritos con un doble propósito. 

<img src=".\img_2\34.png" width="300px" height="300px">


* **Titular**

Incluir un título de diapositiva ayuda a persuadir a la audiencia a prestar atención en ese momento particular de la presentación.

El titular debe resaltar el punto principal de la diapositiva, ser conciso y específico, y ser visible gracias a una fuente más grande.

<img src=".\img_2\35.png" width="300px" height="300px">


* **Método de Superposición**


El Método de Superposición es una poderosa herramientas que rompe una diapositiva compleja en puntos más pequeños.

La idea es tener una diapositiva más limpia y mejor estructurada que facilite el enfoque en cada parte individualmente.

<img src=".\img_2\36.png" width="300px" height="300px">


<img src=".\img_2\37.png" width="300px" height="300px">

Cada punto se presenta y explica por separado, antes de mostrarse juntos para concluir el mensaje.


<img src=".\img_2\38.png" width="300px" height="300px">

Este enfoque ayuda a centrar la atención de nuestra audiencia.


## Visualización de diapositivas


* **Reemplazar muchas frases**

Cuando nos encontremos usando demasiado texto, deberíamos pensar en reemplazarlo con un gráfico. La gente suele entender mucho mejor la información visual. 

<img src=".\img_2\39.png" width="500px" height="300px">


* **Utiliza la superposición y el resaltado**

La superposición y el resaltado también funcionan bastante bien: podemos mostrar cada gráfico uno por uno, y júntelos para resumir la idea. 

Resaltar el gráfico o parte del gráfico también puede ayudarnos a dirigir a nuestra audiencia a lo que estamos explicando. También funciona para texto, para resaltar una idea o una métrica.

<img src=".\img_2\40.png" width="300px" height="300px">


Si es necesario, los titulos también deben usarse siguiendo las mismas pautas que vimos para las diapositivas de texto.


* **Uno o dos graficos de tamaño completo por diapositiva**


Finalmente, debemos usar solo uno o dos gráficos de tamaño completo por diapositiva, previniéndonos de transmitir más de un mensaje por diapositiva, y evitar que saturemos las diapositivas.

# Entrega de la presentación


## Resumen

Ahora entendemos que necesitamos planificar la presentación antes de construir las diapositivas. Y las diapositivas deben construirse para apoyar la historia que queremos transmitir.


## Practica

Ahora, es el momento de discutir la entrega. Una de las cosas más importates es practicar. Escribir un giuón es útil. La idea no es memorizarlo de memoria, sino usarlo como una guia para articular eficientemente los razonamientos y puntos de vista.

Cuando más practiquemos, más familiarizados nos volveremos con nuestro contenido. Además, debemos anticiparnos a las preguntas de seguimiento y pensar en las respuestas adecuadas. Practicar significa ensayar lo más cerca posible del contexto de presentación posible, que implica ponerse de pie, utilizando las diapositivas reales, y hablando en voz alta.

A medida que practicamos, podemos identificar patrones que distraen o palabras de relleno como "um", "así" o "me gusta".

Además, a veces las cosas no suenan tan bien cuando se hablan como cuando se escriben.

También es una oportunidad para encontrar declaraciones de enlace y oraciones para hacer la transición de una diapositiva a la otra.

Por ultimo, pero no menos importante, también debemos practicar cómo responder preguntas de seguimiento que hemos anticipado.


## Realizar la presentación

Practicar nuestras presentación muchas veces es una forma efectiva de familiatizarse con nuestro contenido. Pero hay un par de cosas que debemos considerar al hacer la presentación.

Lo primero es ser conscientes de nuestras emociones. El público se fijara en ellos. Si estamos seguros de nosotros mismo, van a tener confianza en nuestro contenido. Si nos vemos nerviosos o confundidos, es posible que no estén convencidos.


La duración de la capacidad de atención de la audiencia varía según el entorno, pero los investigadores están de acuerdo en que es entre 5 y 20 minutos. Por lo tanto, debemos hacer un esfuerzo adicional para mantener a la audiencia comprometida.

Podemos usar el contacto visual, la interactividad, las preguntas: la idea es simplemente hablarles, no a ellos.

Estamos construyendo una relación con nuestra audiencia: mostremos empatía, actuemos como una guía. No deberíamos usar oraciones como "como sabes" u "obviamente". Podríamos dañar nuestra conexión con la audiencia si sienten que no entienden algo, o si se sienten incómodos haciendo preguntas. 

Debemos asegurarnos de que nuestra presentación no dure más del tiempo asignado. La audiencia se impancientará si tardamos más de lo esperado y su atención disminuirá drásticamente.


El público nos entenderá mejor si hacemos pausas. Si queremos transmitir urgencias, está bien hablar un poco más rapido. Pero después de trasmitir la idea, debemos respirar y darle tiempo a la audiencia para pensar.

Finalmente, siempre debemos estar abiertos a preguntas ya sea durante o al final de la presentación. Mostrará que estamos abiertos a la retroalimentación, interesados en lo que nuestra audiencia tiene que decir, y dispuestos a ayudarlos a entender nuestro mensaje.


## Una presentación oral eficaz


Entonces, en general, para hacer una buena presentación oral siempre debemos recordar:

- Para hablarle a la audiencia (y no a ella)

- Que menos es más

- El mensaje que se lleven a casa debe ser consistente y persistente.

- Practicar  

- Prepararse para las preguntas de seguimiento.

# Evitar errores comunes


## Recapitulemos


Con el fin de ofrecer una presentación conveniente, primero necesitamos planificar lo que vamos a presentar (siguiendo unas pautas).

Un paso importante es preparar y practicar nuestro discurso:cuanto más practicamos, mejor nos volvemos. Por último, podemos entregar la presentación, siempre que evitemos algunos errores comunes.


## Largo

Muchas presentaciones son ineficaces porque son demasiado largas. La audiencia tiene una capacidad de atención que debemos tener en cuenta, y si pasamos de los 20 minutos, corremos el riesgo de perderlos. Además, es importante reservar algo de tiempo para las preguntas que pueda tener la audiencia. Las personas que estén interesadas en nuestro proyecto seguramente tendrán algunas. 


## Proposito


Un error común es olvidar indicar el propósito de la presentación en la introducción. Este simple erro podría arruinar el resto de la presentación, impidiendo que la audiencia entienda la importancia de nuestros hallazgos y dejar en la oscuridad a lo largo de la presentación, disminuyendo el impacto de nuestra historia.


## Orientar al público

Como oradores, debemos guiar a la audiencia, entregar una secuencia de información que apoya nuestra historia, y mantener la atención de nuestra audiencia. Con ese fin, no debemos esperar hasta el final de la presentación para mostrar todos nuestros hallazgos juntos. En su lugar, debemos incluirlos a lo largo de la presentación, proporcionando una columna vertebral al mensaje.



## Participación del público

Otro error común es simplemente dar la presentación como un monólogo, sin comprometer e involucrar a la audiencia. Recuerde que nos estamos presentando ante ellos, no a ellos.


Primero, debemos presentarnos usando una declaración fuerte, diciendo quiénes somos y por qué estamos presentando.

Debido a que nuestra audiencia no es consciente de nuestro proceso de pensamiento, siempre debemos establecer nuestras suposiciones clave. Este gesto demuestra empatía. 

Otra forma de conectarse con la audiencia es hacer preguntas en medio de la presentación. Idealmente, deberíamos saber la respuesta y puede servir como gancho para nuestra siguiente diapositiva.

Por último, podemos mantener a nuestra audiencia comprometica relacionando siempre lo que mostramos con la idea principal. De esa manera, pueden hacer un seguimiento de nuestra historia fácilmente.



## Lenguaje corporal


Si lo que importa es el mensaje, el hablante está en el centro de la presentación. Podemos usar eso a nuestro favor y crear énfasis usando gestos y movimientos naturales: Por ejemplo, moviendo nuestras manos o apuntando a la diapositiva, sonriendo o haciendo una expresión facial. 

Un buen lenguaje corporal transmite confianza y potencia la atención, así nuestra audiencia estará más enfocada en nuestro mensaje.

Esto es especialmente cierto para las presentaciones en vivo, un poco menos para las presentaciones en línea.


## Tonalidad de la voz


Nuestra voz también es una herramienta que debemos utilizar como oradores. Debemos utilizar diferentes tonalidades de voz para enfatizar nuestro mensaje. Por ejemplo, variar la velocidad de nuestra voz hace que nuestra charla sea más interesante.

Podemos hablar rápido para denotar un sentido de urgencia, excitación o emoción, o lento para resaltar la importancia de un hallazgo, introducir nuevas ideas o martillas las fundamentales.


También debemos hablar lo suficientemente alto para que la audiencia pueda escucharnos si estamos dando una presentación en persona. En caso de presentaciones en línea, debemos verificar que nuestro micrófono esté correctamente configurado para que nuestra voz sea clara.


La entonación también importa. Es difícil dar consejos generales, ya que depende del idioma, pero básicamente necesitamos sonar comprometidos. Si no parecemos interessados por lo que presentamos, no podemos esperar que nuestra audiencia lo esté.
