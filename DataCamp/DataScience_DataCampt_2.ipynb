## ¿Qué es la narración de datos?

Por eso, contar historias con datos es primordial. Entonces, que es la narración de datos?

Es un mecanismo poderoso para compartir información respaldada por una narrativa convincente y visualizaciones eficientes.

Las anécdotas impulsan la imaginación de las personas y las historias son más memorables que las métricas y agregan valor al dar contexto a los datos. Gracias a eso, la audiencia puede entender ideas. Así se capta su atención.

Por ejemplo, es posible que tengamos que explicar por qué las predicciones de baja precisión afectan el resultado comercial para los especialistas de la cadena de suministro y cómo podría ayudar una mejor entrada de datos por su parte.

## Narración convincente

## Estructura Narrativa


Para comenzar nuestra historia de datos, debemos mencionar los antecedentes:

   - Detalles sobre qué problemas motivó el análisi.
   - Lo que cambió en la situación previa que ameritaba un análisis.
   - Y en quién se centra el análisis.
       - ¿Estamos analizando los datos de los clientes?, De los empleados?. o de algo más?
       
       
En nuestro caso, mencionamos que frente a las expectativas de la gerencia las utilidades totales de la compañia han disminuido en los últimos tres trimestres.


Después de que se introdujo el problema, debe proporcionar evidencia de los factores que contribuyeron al problema. Pero solo debemos incluir información relevante y no datos detallados que puedan sobrecargar a la audiencia.

Analizamos los datos y descubrimos que las ventas de "chips" aumentaron un 20%. Sin embargo, las ventas de artículos dulces han disminuido en general en un 30%.

## Construyendo narrativa

Al ejecutar un análisis, hay varias cosas que puede observar, que a su vez impulsarán su narrativa.

**Cambios a lo largo del tiempo:** Podemos explorar cómo cambia una característica con el tiempo o el ciclo de tiempo patrones, por ejemplo, cómo las ventas de chocolate con más bajas en verano pero más altas en invierno.

**Correlación:** Podríamos centrarnos en cómo dos cosas se relacionan entre sí, por ejemplo el precio del chocolate y la calificacion del clientes. 

**Comparación:** o mostrar similitudes y diferencias entre el consumo de chocolate entre adultos y niños.

**Custering:** Además, podemos encontrar grupos en los datos, como altos consumidores de chocolate y altos consumidores de café vs. bajos consumidores de chocolares y bajos consumidores de café.


A medida que elaboramos nuestra narrativa, debemos asegurarnos de que estos puntos de datos hagan una historia relevante y reconocible.

## Agregados


- Valor representativo:
    
    - Totales / recuentos
    - Medias
    - Mediana

- La media puede ser engañosa (Outlier)

- Distribucion de los datos

A veces necesitamos resumir datos numéricos en un agregado: un número que da una idea de un conjunto o valor representativo.

Puede ser un simple total o recuento, como las ventas totales o cuánto durará una campaña.

Otro agregado común es la media. El número promedio de chocolares o chips vendidados por año, como vemos en el gráfico.

<img src=".\img_2\15.png" width="300px" height="300px">

O el precio medio de cada producto.

<img src=".\img_2\16.png" width="300px" height="300px">

La media puede ser engañosa, especialmente si hay valores atípicos (valores extremos, los datos no se distribuyen normalmente).
En estos casos, la mediana es una métrica más robusta.

Por ejemplo, el salario anual primedio en 2019 en EE.UU. fue de 51916,27 USD, y salario media anual fue de 34248,45 USD. Usando la media, pensariamos que la tendencia común es un salario de 52K, cuando en realidad la mitad de la población está por debajo de $34K.


## p-value

- Qué es p-value?

    - Valor inferior a 0,05: significación estadística
    - Valor próximo a 0,05: indicador débil

- ¿Qué no es?

    - No es una prueba de evidencia:
        -Rechazar nuestra hipótesis, pero no que sea falsa
    - Considerar métricas alternativas o complementarias
    
    
Al comunicar nuestros resultados, a menudo necesitamos proporcionar pruebas de que son estadísticamente significativos (p, que no pueden debeser a la aleatoriedad).

Un valor de p inferior a 0.05 se considera un indicador de importancia por convención. Cuanto menor sea por debajo  de 0.5 más fuerte es el indicador.

Sin embargo, no es una prueba de prueba: sólo rechaza o acepta una hipótesis sin decir nada acerca de la verdad de la misma.

Debido a la frecuencia con la que las métricas del valor p se malinterpretan o son confusas (la audiencia), considere métricas alternativas o agregue algunas más en apoyo.
