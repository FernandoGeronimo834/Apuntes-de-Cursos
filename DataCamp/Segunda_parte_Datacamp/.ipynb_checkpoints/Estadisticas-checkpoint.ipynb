## Como saber cual medida de centro usar?

La media es mucho más sensible a los valores extremos que la mediana. Dado que la media es más sensible a los valores extremos, funciona mejor para datos simétricos.

Sin embargo, si los datos están sesgados, lo que significa que no son simetricos, como este, generamente es mejor usar la mediana.
<img src=".\img\1.png" width="500px" height="500px">

En este histograma, los datos se apilan a la derecha, con una cola a la izquierda. Los datos que se ven asi se denominan datos asimetricos a la izquierda


<img src=".\img\2.png" width="300px" height="300px">

Cuando los datos se apilan a la izquierda con una cola a la derecha, están sesgados a la derecha.

<img src=".\img\3.png" width="300px" height="300px">

Cuando los datos están sesgados, la media y la mediana son difrentes. 

La media se tira en la direccion del sesgo, por lo que es más baja que la mediana en los datos sesgados a la izquierda y mayor que la mediana en los datos sesgados a la derecha.

Debido a que la media es arrastrada por los valores extremos, es mejor usar la mediana ya que se ve menor afectada por los valores atípicos.

## Que es propagacion?

## Varianza

La varianza, mide el promedio distancia de cada punto de datos a la media de los datos.

<img src=".\img\5.png" width="600px" height="300px">

### Calcular la varianza -> Metodo Manual

1) Empezamos por calcular la distancia entre cada punto y la media, por lo que obtenemos un número para cada punto de dato.

**`dist = df['columna_valores_numericos'] - np.mean(df['columna_valores_numericos'])`**
**`print(dist)`**

2) Luego elevamos al cuadrado cada distancia.

**`sq_dist = dist ** 2`**
**`print(sq_dist)`**

3) Luego sumamos todos esos valores obtenidos:

**`sum_sq_dist = np.sum(sq_dist)`**
**`print(sum_sq_dist)`**

4) Finalmente, dividimos la suma de las distancia al cuadrado por el número de punto de datos menos 1, lo que nos da la varianza.

**`variance = sum_sq_dist / (N°_datos - 1)`**
**`print(variance )`**

## Interpretacion del valor de la varianza:

**cuanto mayor es la varianza, más dispersos están los datos. Hay que tener en cuenta que el valor obtenido esta elevado al cuadrado.**


### Calcular la varianza -> Metódo automatico

**`np.var(df['columna_valores_numericos'], ddof = 1)`**

Si no especificamos que `ddof` es igual a 1. se usa una fórmula ligeramente diferente para calcular la varianza que solo debe usarse en una población completa, no en una muestra.


## Desviacion Estándar

La desviacion estándar es otra medida de dispersión, calculada tomando la raiz cuadrada de la varianza.

**`np.sqrt(np.var(df['columna_valores_numericos'], ddof = 1))`**

Es igual a:

**`np.std(df['columna_valores_numericos'], ddof = 1)`**

Al igual que `np.var`, debemos establecer `ddof` en 1.

Lo bueno de la desviación estándar es que la unidades suelen ser más fáciles de entender ya que no están elevadas al cuadrado.


## Desviacion absoluta media

La desviación absoluta media toma el valor absoluto de la distancia a a la media, y luego toma la media de esas diferencias. Si bien esto es similar a la desviación estándar, no es exactamente lo mismo.

**`dists = df['columns_valores_numericos'] - mean(df['columnas_valors_numericos'])`**
**`np.mean(np.abs(dists)`**

- La desviación estándar eleva al cuadrado las distancia, por lo que las distancias más largas se penalizan más que las más cortas, mientras que la desviación media absoluta penaliza cada distancia por igual.

- Uno no es mejor que el otro, pero SD (desviacion estantar) es mas comun que MAD (Desviacion absoluta media).


## Cuantiles

Los cuantiles, también llamados percentiles, dividen los datos en un cierto número de partes iguales.

**`np.quantile(df['columna_valores_numericos'], 0.5)`**

**$Q_1   \Longrightarrow \ 0.25 $**

**$Q_2  \Longrightarrow \ 0.5 \Longrightarrow Mediana$**

**$Q_3   \Longrightarrow \ 0.75 $**

**Nota: En el lugar donde esta df['columna_de_valores_numericos'] es lugar donde var uan serie de pandas, es decir que puedo tener una serie guardada en una variable, donde dicha serie fue filtrada y aplicada algunos calculos en `.agg` por lo que podria no ocupar el `['columna_valores_numericos']`**

Tambien podemos pasar una lista de números para obtener varios cuantiles a la vez.

**`np.quantile(df['columna_valores_numericos'], [0,0.25,0.5,0.75,1])`**  # dividims los datos en 4 partes iguales

Me devolvera un arreglo con los valores correspondientes a cada cuartil.

Ej: `array([1.9, 7.85, 10.1, 13.75, 19.9])`

Esto significa que el 25% de los datos está entre 1.9 y 7.85, el otro 25% esta entre 7.85 y 10.1, y asi sucesivamente.

donde:

Obtenido el siguiente arreglo: `array([1.9, 7.85, 10.1, 13.75, 19.9])`


## Rango Intercuartil (IQR)

El rango intercuartilico, o IQR, es otra medida de dispersión.

Es la distancia entre los percentiles 25 y 75, que también es la altura de las caja en un diagrama de caja. Podemos calcularlo usando la funcion cuantil o usando la funcion `iqr` de `scipy.stats` para obtener 5.9 horas (con datos del ejemplo).

$IQR = Q_3 - Q_1$

**`np.quantile(df['columna_valores_numericos'], 0.75 - df['columna_valores_numericos'], 0.25)`**

**Tambien podria obtenerlo con la libreria `scipy.stats`**

**`from scipy.stats import iqr`**

**`iqr(df['columna_valores_numericos'])`**



## Outliers (Valores atípicos)

Los valores atípicos son puntos de datos que son sustancialmente diferentes de los demás.

**Pero, Como sabemos qué es una diferencia sustancial?**

Una regla que se usa a menudo es que cualquier punto de datos menor que el primer cuartil menos 1.5 veces el IQR es un valor atípico, asi como cualquier punto mayor que el tercer cuartil más 1.5 veces IQR.

Entonces:

**Valores Atipicos o Outliers**

Es un valor atipico, aquel valor que cumpla cualquiera de estas dos condiciones

- $Valor < Q_1 -1.5 * IQR$ 
- $Valor > Q_3 + 1.5 * IQR$

## Encontrar valores Atípicos

Para encontrar valores atípicos, comenzaremos calculando el IQR.

**`from scipy.stats import iqr`**

**`iqr = iqr(df['columna_valores_numericos'])`**

Luego podemos calculr los umbrales inferior y superior siguiente las formulas.

**`lower_threshold = np.quantile(df['columna_valores_numericos'], 0.25) - 1.5 * iqr`**

**`upper_threshold = np.quantile(df['columna_valores_numericos'], 0.75) + 1.5 * iqr`**

Ahora podemos crear un subconjunto del DataFrame para encontrar los valores que estén por debajo o por encima de los umbrales.

**`df[(df['columna_valores_numericos'] < lower_threshold) | (df['columna_valores_numericos'] > upper_threshold)]`**

La probabilidad siempre está entre cero y 100 porciento. Si la probabilidad de algo es cero, es imposible, y si la probabilidad de algo es del 100%, ciertamente sucederá.

A continuacion con el metodo **`.sample()`** (si coloco un numero entero entre los parentesis obtendre n filas aleatorias), de forma predeterminada, muestra aleatoriamente una fila del DataFrame

## Configurar la semilla aleatoria

Se configura dicha semilla cuando se desea ejecturar un experimento aleatorio y obtener el mismo resultado en todas las ejecuciones.

**`np.random.seed(10)`**

La semilla es un número que el generador de números aleatorios de Python usa como punto de partida, por lo que si lo orientamos con un número semilla, generará el mismo valor aleatorio cada vez. 

El número de la semilla en si no importa. Lo único que importa es que usemos la misma semilla la próxima vez que ejecutemos el script.

**Muestreo sin reemplazo: Es cuando al tomar el nombre de una persona de la caja, al querer sacar otro nombre, el nombre que salio primero no vuelve a la caja al momento de sacar el segundo nombre, por lo que al sacar el segundo nombre el espacio muestral es menor al inicial.**

**Muestreo con reemplazo: Sucede cuando se toma en cuenta un nombre sacado la primera vez, cuando se quiere sacar el segundo nombre, por lo que el mismo nombre puede salir más de una vez, es decir el espacio muestral es el mismo cuando se saca el primer nombre que cuando se saca el segundo nombre.**

## Muestreo con reemplazo

**`df.sample(num_filas_aleatorias, replace = True)`**

Me devolvera un dataframe con 5 filas aleatorias, donde los nombre de las personas de la caja pueden repetirse. El argumento `replace` establecido en `True` es lo que permite el muestreo con reemplazo.


## Eventos Independientes

Dos eventos son independientes si la probabilidad del segundo evento no se ve afectada por el resultado del primero.

Por ejemplo, si estamos muestreando con reemplazo, la probabilidad que Claire sea elegida en segundo lugar es del 25%, sin importar quién sea elegido primero.

En general, al muestrear con reemplazo, cada selección es independiente.

<img src=".\img\9.png" width="300px" height="300px">


## Eventos Dependientes

De manera similar, los eventos se consideran dependientes cuando el resultado del primero cambia la probabilidad del segundo.

Si muestreo sin reemplazo, la probbilidad de que Claire sea elegida en segundo lugar depende de quién sea elegido primero.

Si Claire es elegida primero, hay 0% de probabilidad de que Claire sea elegida en segundo lugar. Si alguien más es elegido primero, hay un 33% de probabilidad de que Claire sea elegida en segundo lugar.

En general, cuando se muestrea sin reemplazo, cada selección es dependiente.

# La distribucion Binomial

El experimento de la moneda donde solo puedo obtener dos resultados, cara o cruz, el cual podriamos tambien llamar exito o fracaso, o 1 y 0. Como se puede observar se obtiene un resultado binario, donde la probabilidad de que salga 1 o 0 es del 50% para ambos casos. Es en estos casos donde se aplica la distribucion Binomial, para resultados binarios, es decir donde solo hay un caso de exito, y uno de fracaso

Esto se puede simular en Python con la libreria `scipy`, haciendo uso de la distribucion binimial de la siguiente forma:

**`binom.rvs(N° de monedas, probabilidad de cara o cruz, size = N° de intentos)`**

La funcion `binom.rvs`, que toma el número de monedas que queremos lanzar, la probabilida de cara o exito, y un argumento llamado tamaño, que es el número de intentos.

`size` es un argumento con nombre, por lo que necesitaremos especificar explícitamente que el tercer argumento corresponde al tamaño, o obtendremos resultados incorrectos.

Esta llamada devolverá un 1, que contaremos como cara, o un 0, que contaremos como cruz.

Podemos usar `binom.rvs 1, 0.5, size = 1 para lanzar 1 monera, con un 50% de probabilidad de cara, 1 vez.`

**`from scipy.stats import binom`**

**`binom.rvs(1, 0.5, size = 1)`**

Resultado --> `array([1])`


Para Realizar ocho lanzamientos de monedas, podemos cambiar el argumento `size` a 8, que lanzará 1 monera con un 50% de probabilidad de cada 8 veces.

## Distribucion Binomial

La distribución Binomial describe la probabilidad del número de éxitos en una secuencia de ensatos independientes.

En otras palabras, puede decirnos la probabilidad de obtener cierto número de caras en una secuencia de lanzamientos de monedas.

Tenga en cuenta que esta es una distribucion discreta ya que estamos trabajando con un resultado contable.

La distribucion binomial puede se describir usando dos parámetros, n y p.

- $n$: representa el número total de ensayos que se están realizando
- $p$: es la probabilidad de éxito.

$n$ y $p$ son también el tercer y segundo argumento de `binom.rvs`

<img src=".\img\32.png" width="300px" height="300px">

Asi es como se ve la distribucion de 10 monedas.

<img src=".\img\33.png" width="300px" height="300px">

Tenemos la mayor posibilidad de obtener 5 caras en total, y una posibilidad mucho menor de obtener 0 caras o 10 caras.

## Probabilidades con la Distribucion Binomial

### Cual es la probabilidad de obtener 7 caras de 10 monedas?

Podemos usar `binom.pmf` donde:

- El primer argumento es el número de caras o éxitos.
- El segundo argumento es el número de intentos, n.
- El tercer argumento es la probabilidad de éxito, p.

**Ejemplo N° 1:**

$P(caras = 7)$

**`binom.pmf(7, 10, 0.5)`**

Posible resultado -> `0.1171875`

si lanzamos 10 monedas, hay un 12% de posibilidades de que exactamente 7 de ellas sean cara.

**Ejemplo N° 2:**

$P(caras <= 7)$

`binom.cdf` da la probabilidad de obtener un número de éxitos menor o igual que el primer argumento.

**`binom.cdf(7,10, 0.5)`**

Posible resultado -> `0.9453125`

La probabilidad de obtener 7 caras o menos de 10 monedas es de alrededor del 95%


**Ejemplo N° 3:**

$P(caras > 7)$

Podemos tomar 1 menos la probabilidad de obtener 7 caras o menos para obtener la probabilidad de un número de éxitos mayor que el primer argumento.

**`1 - binom.cdf(7, 10, 0.5)`**

## Valor esperado

El valor esperado de la distribucion binomial se puede calcular multiplicando $n$ por $p$.

**$Valor \ Esperado = n * p$**

El número esperado de caras que obtendremos al lanzar 10 monedas es 10 veces 0.5, que es 5.

$Valor \ Esperado = 10 * 0.5 = 5$


## Independencia

Es importante recordar que para que se aplique la distribución binomial, la prueba debe ser independiente, por lo que el resultado de una prueba no deberia afectar al siguiente. Por ejemplo, si elegimos al azar de estas tarjetas con ceros y unos, tenemos una probabilidad del 50-50 de obtener un 0 o un 1. Pero como estamos muestreando sin reemplazo, las probabidades para el segundo ensayo son diferentes debido al resultado del primer ensayo.

Dado que estos ensayos no son independientes, no podemos calcular probabilidades precisas para esta situación utilizando la distribucion binomial.

En nuestro ejemplo, la distribucion muestral se acercó más a la distribucion normal a medida que tomamos más y más medias muestrales.

# La Distribucion de Poisson

## Proceso de Poisson

- Un proceso de Poisson es un proceso en el que los eventos parecen suceder a una determina velocidad, pero completamente al azar.

- Por ejemplo: 

    - El número de animales adoptados de un refugio animal cada semana es un proceso de Poisson: Podemos saber que en promedio hay 8 adopciones por semana, pero este número puede diferir aleatoriamente.
    - El número de personas que llegan a un restaurante cada hora.
    - El número de terremotos por ali en California.

- La unidad de tiempo, como horas, semanas o años, es irrelevante siempre que sea consistente.


## Distribucion de Poisson

La distribución de Poisson describe la probabilidad de que ocurra una cierta cantidad de evento durante un período fijo de tiempo.

Podemos usar la distribución de Poisson para calcular la probabilida de al menos 5 animales adoptados en una semana, la probabilidad de que llegen 12 personas en un restaurante, o la probabilida de menos de 20 terremotos en Californa en un año.


### Lambda ($\lambda$)

La distribución de Poisson se describe mediante un valor llamado lambda, que representa el número promedio de eventos por período de tiempo.

En el ejemplo del refugio de animales, este seria el número promedio de adopciones por semana, que es 8. Este valor es tambien el valor esperado de la distribución.

Note que es una distribucion discreta ya que estamos contando eventos, y 7 y 8 son el número más probable de adopciones en una semana.

<img src=".\img\56.png" width="400px" height="300px">


## Lambda es el pico de la distribucion

Lambda cambia la forma de la distribucion, por lo que una distribucion de Poisson con lambda es igual a 1, en azul, se ve bastante diferente a una distribucion de Poisson con lambda igual a 8, en verde, pero pase lo que pase, el pico de la distribución siempre está en su valor lambda.

<img src=".\img\57.png" width="400px" height="300px">

## Probabilidad de un valor unico

Dado que el número promedio de adopciones por sema es 8, Cuál es la probabilidad de 5 adopciones en una semana?

Al igual que las otras distribuciones de probabilidad, podemos importar poisson desde `scipy.stats`.

Usaremos la funcion `possion.pmf`, pasando `5` como el primer argumento y 8 como el segundo argumento para indicar la media de la distribucion.

**`from scipy.stats import poisson`**

**`poisson.pmf(5,8)`**

Resultado -> `0.09160366`  # Casi 9%

## Probabilidad menor o igual de:

Para obtener la probabilida de que ocurran 5 adopciones o menos en una semana, use la funcion `poisson.cdf` los mismo números.

**`from scipy.stats import poisson`**

**`poisson.cdf(5,8)`**

Resultado --> `0.1912361`  # Alrededir del 20%

## Probabilidad mayor que:

Al igual que otras fucniones de probabilidad que has aprendido hasta ahora, toma 1 menos la probabilidad "menor o igual a 5"
para obtener la probabilidad de más de 5 adopciones.

**`1 - poisson.cdf(5, 8)`**

Resultado -> `0.8087639` # Hay un 81% de posibilidad de que se produzcan más de 5 adopciones.


Si el promedio de adopciones aumenta a 10 por semana.

**`1 - poisson.cdf(5, 10)`**

Resultado -> `0.932914` # habra un 93% de posibilidad de que ocurran más de 5 adopciones.


## Muestreo desde una distribucion de Poisson

Al igual que otras idstribucinoes, podemos tomar muestras de las idstribuciones de Poisson usando `poisson.rvs`.

Aqui, simularemos 10 semanas diferentes en el refugio de animales. 

**`from scipy.stats import poisson`**

**`poisson.rvs(8, size = 10)`**

Resultado -> `array([9, 9, 8, 7, 11, 3, 10, 6, 8, 14])`

En una semana hay 14 adopciones, pero en otra solo 6.

# Más Distribucinos de Probabilidad

## Distribucion Exponencial

Representa la probabilidad de que pase cierto tiempo entre eventos de Poisson.

Podemos usar la distribucion exponencial para predecir, por ejemplo: 
- La probabilidad de más de 1 dia entre adopciones.
- La probabilidad de menos de 10 minutos entre llegadas a restaurantes.
- La probabilidad de que pasen 6-8 meses entre terremotos.

Al igual que la distribucion de Poisson, la unidad de tiempo no importa siempre que sea consistente.

La distribución exponencial utiliza el mismo valor de lambda, que representa la tasa, que lo hace la distribucion de Poisson. Tenga en cuenta que lambda y rate significan el mismo valor en este contexto.

Tambien es continua, a diferencia de la distribucion de Poisson, ya que representa el tiempo.


#### Ejemplo: Solicitudes de atencion al cliente.


Digamos que se crea un ticket de servicio al cliente cada 2 minutos. Podemos reformular esto para que sea en términos de un intervalo de tiempo de un minuto, por lo que se crea la mitad de un ticket cada minuto, entonces usaremos 0.5 como valor de lambda.

$\lambda = 0.5$

La distribucion exponencial con una tasa de un medio se ve asi:

<img src=".\img\59.png" width="500px" height="300px">

## Lamnda en la distribucion exponencial

La tasa afecta la forma de la distribución y qué tan abruptamente declina.

<img src=".\img\60.png" width="500px" height="300px">


## Valor esperado de la distribucion exponencial

Recuerde que lmabda es el valor esperada de la Distribución de Poisson, que mide la frecuencia en términos de tasa o números de eventos.

En nuestro ejemplo del ticket de servicio al cliente, esto significa que la cantidad esperada de solicitudes por minuto es de 0.5.

La distribucion exponencial mide la frecuencia en términos de tiempo entre eventos.

El valor esperado de la distribución exponencial se puede calcular tomando:

**$1 / \lambda$**

En nuestro ejemplo, el tiempo esperado entre solicitudes es 1 sobre la mitad, que es 1, por lo que hay un promedio de 2 minutos entre solicitudes.

## Distribución Exponencial en Python

De manera similar a otras distribuciones continuas, podemos usar `expon.cdf` para calcular probabilidades.

### Ejemplo 1:  $P(Esperar < 1 min) =?$

La probabilidad de esperar menos de 1 minutos por una nueva solicitud se calcula usando `expon.cdf`, pasar 1 seguido de escala es igual al punto 5.

**`from scipy.stats import expon`**

**`expon.cdf(1, scale = 0.5)`**

Resultado -> `0.864664716...` -> 86%


### Ejemplo 2: $P(Esperar > 3min) = ?$

La probabilidad de esperar más de 3 minutos se puede encontrar usando:

**`from scipy.stats import expon`**

**`1 - expon.cdf(3, scale = 0.5)`**

Resultado -> `0.00247875...` -> 0.2%.

### Ejemplo 3: $P(1min < Esperar < 3min) = ?$

La probabilidad de esperar entre 1 y 3 minutos se pueden encontrar de la siguiente manera:

**`from scipy.stats import expon`**

**`expon.cdf(3, scale = 0.5) - expon.cdf(1, scale = 0.5)`**

Resultado -> `0.1328...` -> 13%

Hay un 13% de posibilidades de que espere entre 1 y 3 minutos.

## Otras Transformaciones

Además de la transformación logaritmica ($log(x)$), hay muchas otras transformaciones que se puede usar para hacer una relación más lineal, como sacar la raiz cuadrada ($sqrt(x)$) o el reciproco de una variable ($1/x$).

La elección de la transformación dependerá de los datos y de cuán sesgados estén.

Estos se pueden aplicar en diferentes combinaciones a x e y, por ejemplo, podria aplicar una transformacion tanto a x como a y, o una transformación de raiz cuadrada a x y una transformacion reciproca a y.

## Porque usar una transformación?

Ciertos métodos estadísticos se basan en variables que tienen una relación lineal, como calcular un coeficiente de correlación. La regresión lineal es otra técnica estadística que requiere relacionarse variables de manera linea.

## La correlación no implica causalidad

Una advertencia más importante de la correlación que es posible que hayas oido hablar antes: la correlación no implica causalidad.

Esto significa que si x e y están correlacionados, x no necesariamente causa y.

Por ejemplo, aqui hay un diagrama de dispersion de la margarita per cápita consumo de los EE.UU. cada año y la tasa de divorcios en el estado de Maine.

La correlación entre estas dos variables es 0.99, que es casi perfecta. Sin embargo, esto no significa que consumir más margaritas provoque más divorcios.

Este tipo de correlación a menudo se denomina correlación espuria.

## Confusión

Un fenómeno llamado confusión puede dar lugar a correlaciones espurias. Digamos que queremos saber si beber café causa cáncer de pulmón. Mirando los datos, encontramos que el consumo de café y el cáncer de pulmón son correlacionados, lo que puede llevarnos a pensar que beber más café le dará cáncer de pulmón. Sin embargo, hay una tercera variable oculta en juego, que es fumar. Se sabe que fumar está asociado con el consumo de café. Tambien se sabe que fumar causa cáncer de pulmón. En realidad, resulta que el café no causa cáncer de pulmón y solo se asocia con él, pero parece causal debido a la tercera variable, el tabaquismo.

Esta tercera variable se denomina factor de confusión o variable oculta. Esto significa que la relación de interés entre el café y el cáncer de pulmón es una correlacion espuria.


Otro ejemplo de esto es la relación entre las vacaciones y las ventas minoristas. Si bien puede ser que la gente compre más durante las festividaddes como una forma de celebración, es dificil decir cuánto de el aumento de las ventas se debe a las vacaciones, y cuánto se debe a las ofertas y promociones especiales que a menudo se realizan durante las vacaciones. 

Aqui, las ofertas especiales confunden la relación entre las vacaciones y las ventas.

# Diseño de Experimentos


## Vocabulario

Los expremientos generalmente tienen como objetivo responder una pregunta en la forma "¿Cuál es el efecto del tratamiento en la respuesta?"

En este escenario, el tratamiento se refiere a lo explicativo o variable independiente, y respuesta se refiere a la respuesta o variable dependiente. Por ejemplo, cuál es el efecto de un anuncio en la cantidad de productos comprados?.

En este caso, el tratamiento es un anuncio y la respuesta es la cantidad de productos comprados.

## Experimentos controlados

En un experimento controlado, los participantes se asignan al azar al grupo de tratamiento o el grupo de control, donde el grupo de tratamiento redibe el tratamiento y el grupo de control no. Un gran ejemplo de esto es una prueba A/B.

En nuestro ejemplo, el grupo de tratamiento verá un anuncio y el grupo de control no. Aparte de esta diferencia, los grupos debe ser comparable para que podamos determinar si ver un anucio hace que la gente compre más. Si los grupos no son comparables, esto podría generar confución o sesgo.

Si la edad promedio de los participantes en el grupo de tratamiento es de 25 años y la edad promedio de participantes en el grupo de control tiene 50 años, la edad podria ser un posible factor de confusíon si es más joven es más probable que las personas compren más, y esto hará que el experimento sesgue hacia el tratamiento.

## El estandar de oro de experimentos utilizará

El estándar de oro, o experimento ideal, eliminará tanto sesgo como sea posible, median el uso de ciertas herramientas.

La primera herramienta para ayudar a eliminar el sesgo en los experimentos controlados es utilizar un ensayo controlado aleatorio. En un ensayo controlado aleatorio, los participantes se asignan al azar el tratamiento o grupo de control y su asignación no se basa en otra cosa que no sea el azar.

La asignacion aleatoria como esta ayuda a garantizar que los grupos sean comparables.

La segunda forma es usar un placebo, que es algo que se parece al tratamiento, pero no tiene efecto. De esta manera, los participante no saben si están en el grupo de tratamiento o de control. Esto asegura que el efecto del tratamiento se deba al tratamiento en si mismo, no a la idea de recibir el tratamiento.

Esto es común en los ensayos clínicos que prueban la eficacion d eun fármaco. El grupo de control aún recibirá una pastilla, pero es una pastilla de azúcar que tiene efectos mínimos en la respuesta.


En un experimento doble ciego, la persona que administra el tratamiento o ejecuta el experimento tampoco sabe si están administrando el tratamiento real o el placebo. Esto protege contra el sesgo en la respuetsa, así como el análisis de los rsultados.

Todas estas diferentes herramientas se reducen al mismo principio: Cuantas menos oportunidades hata para que el sesgo se infiltre en su experimento, más fiable podrá concluir si el tratamiento efecto la respuesta.


## Estudio Observacional

En un estudio de observacion, los participantes no se asignan aleatoriamente a los grupos. En cambio, los participantes se asignan a si mismos, generamenteen función de caracteristicas preexistentes.

Esto es útil para responder preguntas que no conducen a un experimento controlado. Si desea estudiar el efecto de fumar sobre el cáncer, no puede obligar a las personas a comenzar a fumar. De manera similar, si desea estudiar cómo el comportamiento de compras anterior afecta si alguien comprará un producto, no se puede oblicar a la gente a tener cierto comportamiento de compra en el pasado.

Debido a que la asignación no es aleatoria, no hay forma de garantizar que los grupos serán comparables en todos los aspectos, por lo que los estudios observacionales no pueden establecer causalidad, solo asociación.

Los efectos del tratamiento pueden ser confundidos por factores que llevarons a ciertas personas al grupo de control y a ciertas personas al grupo de tratamiento. Sin embargo, hay formas de controlar los factores de confusión, que pueden ayudar a fortalecer la confiabilidad de las conclusiones acerca de la asociación.


## Estudios longitudinales frente a estudios transversales

La última distinción importante a hacer es entre estudios longitudinales y transversales.

### Estudio Longitudinal

- Sigue a los participantes durante un período de tiempo para examinar el efecto del tratamiento en la respuesta.

- El efecto de la edad en la estatura no se confunde con la generación

- Más caro, el resultado tarda más tiempo

### Estudio Transversal

- Los datos se recopilan a partir de una única instantánea en el tiempo

- El efecto de la edad sobre la altura se ve afectado por la generación

- Más baratos, rapidos y más convenientes.

Si quisiera investigar el efecto de la edad en altura, un estudio transversal mediría las alturas de personas de diferentes edades y las compararía. Sin embargo, los resultados se confundirán por año de nacimiento y estilo de vida ya que es posible que cada generación sea más alta.

En un estudio longitudinal, las mismas personas tendrían sus alturas registrados en diferentes momentos de sus vidas, por lo que se elimina la confusión.
