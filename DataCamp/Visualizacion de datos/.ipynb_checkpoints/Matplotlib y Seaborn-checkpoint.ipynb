## Introduccion a visualizacion de datos con Matplotlib

**`import matplotlib.pyplot as plt`**

El objeto Figure es un contenedor que contiene todo lo que ves en la página. Mientras tanto, Axes es la parte de la página que contiene los datos.

**`fig, ax = plt.subplots()`**

Si el lienzo sobre el que dibujaremos con nuestros datos, para visualizarlos.

**`plt.show()`**

**`ax.plot(seattle_weather['month'], seattle_weather['mly-tavg-normal'])`**

**`plt.show()`**

Me devolvera una curva donde cada punto se unira entre si. Eje x: Meses, Eje y: temperatura de la ciudad.

# Pequeños Múltiplos (subparcelas)

Estas son varias parcelas pequeñas que muestran datos similares en diferentes condiciones.

**`fig, ax = plt.subplots(numero_de_filas, numero_de_columnas)`**

**`plt.show()`**

Lo que se creara con este codigo, es un objeto Figura con tres filas de subgraficos y dos columnas.

## Agregar datos a subparcelas

**`ax.shape`** --> (numero_de_columnas, numeros_de_filas)

Para agregar datos, ahora tendriamos que index este objeto y llamar al método de trazado en un elemento de la matriz.

**`ax[0, 0].plot(seattle_weather.plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-NORMAL'], color = 'b')`**

**`plt.show`**

Me mostrara la grafica en la subparcela superior izquierda.

## Subparcelas con datos

Existe un caso especial para situaciones en las que solo tiene una fila o solo una columna de gráficos. En este caso, la matriz resultando será unidimensional y usted solo tendrá que proporcionar un indice para acceder a los elemento de esta matriz.

**`fig, ax = plt.subplots(2, 1)`**

**`ax[0].plot(seattle_weather.plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-NORMAL'], color = 'b')`**

**`ax[1].plot(austin_weather.plot(seattle_weather['MONTH'], austin_weather['MLY-PRCP-NORMAL'], color = 'b')`**

**`ax[0].set_ylabel('Etiqueta_eje_y')`**

**`ax[1].set_ylabel('Etiqueta_eje_y')`**

**`ax[1].set_xlabel('Etiqueta_eje_x')`**


## Compartiene el rango del eje Y

Es posible que al graficar diferentes columnas de datos, los valores esten alejados en cuando el valor entre columnas, por lo que si graficamos los datos en diferentes parcelas, es posible que el rango del eje y no sean iguales entre las subparcelas, lo que nos podria afectar al momento de desear comparar las grafica, es por ello que se puede configurar el eje Y, de forma que ambas graficas tengan el mismo rango, y lo haremos en el argumento `sharey`.

**`fig, ax = plt.subplots(numero_de_filas, numero_de_columnas, sharey = True)`**

Las graficas tendra el mismo rango en el eje Y.

# Trazado de series de tiempo con diferentes variables

## Trazar dos serie temporales juntas

Podemos quere trazar en una misma figura, dos graficas, donde el eje x, es la variable de tiempo, y en el eje y, querer graficar dos variables diferentes. Pero surge un problema al momento de graficar, y es que la escala de las variables es diferente, lo cual es posible que las graficas no se muestren la manera más optima.

```python
fig, ax = plt.subplots()
ax.plot(clima.index, clima['co2'])
ax.plot(clima.index, clima['relative_temp'])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')
plt.show()
```

## Usando ejes gemelos

Vamos a trazar ambas graficas en la misma subparcela, usando dos escalas del eje Y diferentes.

Comenzamos agregando la primera variable a nuestros ejes. Luego , usamod el método `twinx` para crear un gemelo de estos ejes.

Esto significa que los dos ejes comparten el mismo eje x, pero los ejes Y están separados.

Agregamos la otra variable a este segundo objeto Axes y mostramos la figura.

```python
fig, ax = plt.subplots()
ax.plot(clima.index, clima['co2'], alpha = 0.9)
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)', color ='b')

ax2 = ax.twinx()
ax2.plot(clima.index, clima['relative_temp'], color = 'r', alpha= 0.4)
ax2.set_ylabel('Relative temperature (Celsius)', color ='r')

plt.show()
```

## Color de etiquetas y color de ticks

Podemos modificar el color de las etiquetas para que sea más claro a que grafica pertenece cada escala, esto lo podemos hacer agregando el argumento `color` en los `set_ylabel`.

Tambien podemos modificar el color de los ticks, agregando un par de lineas de codigo con el metodo `ticks_params` donde el primer argumento, puede ser `x` o `y`, y para modificar el color agrego el argumento `colors`.

```python
fig, ax = plt.subplots()
ax.plot(clima.index, clima['co2'], alpha = 0.9)
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)', color ='b')
ax.tick_params('y', colors = 'b')

ax2 = ax.twinx()
ax2.plot(clima.index, clima['relative_temp'], color = 'r')
ax2.set_ylabel('Relative temperature (Celsius)', color ='r')
ax2.tick_params('y', colors = 'r')

plt.show()
```

# Una funcion que traza series temporales


```python
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color = color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color = color)
    axes.tick_params('y', colors = color)
```

```python
# USANDO LA FUNCION

fig, ax = plt.subplots()

plot_timeseries(ax, clima.index, clima['co2'], 'blue', 'Time','CO2 (ppm)')

ax2 = ax.twinx()

plot_timeseries(ax2, clima.index, clima['relative_temp'], 'red','Time','Relative temperature (Celsius)')

plt.show()
```

# Anotar datos en series temporales

Las anotaciones suelens er pequeños fragmento de texto que se refieren a una parte particular de la visualización, enfocando nuestra atención en alguna caracteristica de los datos explicando esta cracterística.

Una forma de llamar la atención sobre una parte de una trama es anotarla.

## Anotación

Esto significa dibujar una flecha que apunte a una parte de la trama y podes incluir texto para explicarla.

Por ejemplo, digamos que notamos que la primera fecha en la que la temperatura relativa superó 1 grado celcius fue el 6 de octubre de 2015.

Nos gustaria señalar esto en la trama. Para hacer esto usamos el metodo `.annotate('texto_descriptivo', xy = (coordenada de dodne apuntara la flecha ej: pd.Timestamp('2015-10-06'), 1))`

```python
fig, ax = plt.subplots()

plot_timeseries(ax, clima.index, clima['co2'], 'blue', 'Time','CO2 (ppm)')

ax2 = ax.twinx()

plot_timeseries(ax2, clima.index, clima['relative_temp'], 'red','Time','Relative temperature (Celsius)')

ax2.annotate('>1 degree', xy= (pd.Timestamp('2015-10-06'), 1))

plt.show()
```

## Posicionar texto

El método de anotación toma un argumento de texto xy opcional que selecciona la posición xy del texto. La coordenada de posicionamiento del texto se logra experimentando con dichas coordenadas, para hallar la mejor posicion. Luego se vera la forma de indicar a que parte de la trama corresponde dicho texto.

`ax2.annotate('>1 degree', xy= (pd.Timestamp('2015-10-06'), 1),
                xytext = (pd.Timestamp('2008-10-06'),-0.2))`

```python
fig, ax = plt.subplots()

plot_timeseries(ax, clima.index, clima['co2'], 'blue', 'Time','CO2 (ppm)')

ax2 = ax.twinx()

plot_timeseries(ax2, clima.index, clima['relative_temp'], 'red','Time','Relative temperature (Celsius)')

ax2.annotate('>1 degree', xy= (pd.Timestamp('2015-10-06'), 1), xytext = (pd.Timestamp('2008-10-06'),-0.2))

plt.show()
```

## Agregar flecha a la anotacion

Para conectar entre el texto de la anotación y los datos anotados, podemos agregar una flecha.

El argumento de palabra clave para hacer esto se llama `arrowprops`, que significa propiedades de flecha. Este argumento de palabra clave toma como entrada un diccionario que define las propiedades de la flecha que nos gustaria usar. Si pasamos un diccionario vacío al argumento de la palabra clave, la flecha tendrá las propiedades predeterminadas, como se muestra acontinuacion.

```python
fig, ax = plt.subplots()

plot_timeseries(ax, clima.index, clima['co2'], 'blue', 'Time','CO2 (ppm)')

ax2 = ax.twinx()

plot_timeseries(ax2, clima.index, clima['relative_temp'], 'red','Time','Relative temperature (Celsius)')

ax2.annotate('>1 degree', xy= (pd.Timestamp('2015-10-06'), 1), xytext = (pd.Timestamp('2008-10-06'),-0.2), arrowprops = {})

plt.show()
```

## Personalizar propiedades de la flecha

Podemos personalizar la apariencia de la flecha.

```python
fig, ax = plt.subplots()

plot_timeseries(ax, clima.index, clima['co2'], 'blue', 'Time','CO2 (ppm)')

ax2 = ax.twinx()

plot_timeseries(ax2, clima.index, clima['relative_temp'], 'red','Time','Relative temperature (Celsius)')

ax2.annotate('>1 degree', xy= (pd.Timestamp('2015-10-06'), 1), xytext = (pd.Timestamp('2008-10-06'),-0.2), 
             arrowprops = {'arrowstyle':'->', 'color':'gray'})

plt.show()
```

# Comparacion Cuantitativa: Histogramas

Los histogramas: Esta visualizacion es útil porque nos muestra la distribución completa de valores dentro de una variable.

## Grafico de barras

```python
summer2016_nadadores = summer2016[(summer2016['Sex'] == 'M') & (summer2016['Sport'] == 'Rowing')]

mens_gymnastics = summer2016[(summer2016['Sex'] == 'M') & (summer2016['Sport'] == 'Gymnastics')]
summer2016_nadadores.head()
```

```python
fig, ax = plt.subplots()

ax.bar('Rowing', summer2016_nadadores['Height'].mean())
ax.bar('Gymnastics', mens_gymnastics['Height'].mean())
ax.set_ylabel('Height (cm)')
```

# Introduccion a Histogramas

Pero, en cambio, un histograma mostraría la distribución completa de valores dentro de cada variable.

```python
fig, ax = plt.subplots()

ax.hist(summer2016_nadadores['Height'])
ax.hist(mens_gymnastics['Height'])

ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of observations')

plt.show()
```

En el histograma que se muestra, el eje x son los valores dentro de la variable y la altura de las barras representa el número de observaciones dentro de un contenedor particular de valores.

Por ejemplo, hay 12 gimnastas con alturas entre 164 y 167 cm, por lo que la barra más alta del histograma naranja tiene 12 unidades de altura.

Asi mismo, hay 20 remeros con alturas entre 188 y 192 cm, y la barra más alta del histograma azul tiene 20 unidades de altura.

## Las etiquetas son necesarias

Debido a que la etiqueta del eje x ya no proporcionan información sobre el qué el color representa qué variable, las etiquetas son realmente necesarias en los histogramas. Es por ello que en el metodo `hist` agregaremos el argumento `label` Para agregar una etiqueta al histograma, y luego agregar `ax.legend()`

```python
fig, ax = plt.subplots()

ax.hist(summer2016_nadadores['Height'], label = 'Rowing')
ax.hist(mens_gymnastics['Height'], label = 'Gymnastics')

ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of observations')

ax.legend()

plt.show()
```

## Personalizar histograma: Configurar el numero de contenedores

Por defecto, el número de barras o cotenedores en un histograma es 10, pero podemo personalizarlo. Si proporcionamos el número entero al argumento de palabras clave `bins`, el histograma tendrá ese número de contenedores.

```python
fig, ax = plt.subplots()

ax.hist(summer2016_nadadores['Height'], label = 'Rowing', bins = 5)
ax.hist(mens_gymnastics['Height'], label = 'Gymnastics', bins = 5)

ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of observations')

ax.legend()

plt.show()
```

## Personalizar Histograma: Establecer los límites del contenedor

Si, en cambio, proporcionamos una secuencia de valores, estos números se establecerá como los límites entre los contenedores, como se muestra acontinuacion.

```python
fig, ax = plt.subplots()

ax.hist(summer2016_nadadores['Height'], label = 'Rowing', bins = [150, 160, 170, 180, 190, 200, 210])
ax.hist(mens_gymnastics['Height'], label = 'Gymnastics', bins = [150, 160, 170, 180, 190, 200, 210])

ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of observations')

ax.legend()

plt.show()
```

## Personalizar Histograma: Transparencia

Al mirar al enterior histograma, podría preguntame si hay cualquier medallista de remo con una latura inferior a  180 centimetros. Esto es dificil de decir porque las barras del historgrama de gimnasia ocluyen esta información.

La oclusión se puede eliminar cambiando el tipo de histograma que se utiliza (`histtype`). En lugar del tipo de "barra" que se utiliza de forma predeterminada, puede especificar un tipo de historial de "paso", que muestra el histograma con lpineas finas, en lugar de sólidas.


Exponiendo que si: hay remeros con una altura de menosd e 180 centimetros.

```python
fig, ax = plt.subplots()

ax.hist(summer2016_nadadores['Height'], label = 'Rowing', 
        bins = [150, 160, 170, 180, 190, 200, 210],
       histtype = 'step')
ax.hist(mens_gymnastics['Height'], label = 'Gymnastics', 
        bins = [150, 160, 170, 180, 190, 200, 210],
       histtype = 'step')

ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of observations')

ax.legend()

plt.show()
```

# Trazado estadistico

El trazado estadístico es un conjunto de método para utilizar la visualización para hacer comparaciones.

## Agregar barras de error a los grafica de barras: Metodo 1

Estos son marcadores adicionales en una gráfica o gráfico de barras que nos dicen algo sobre la distribucion de los datos.

Los histogramas, que se vieron anteriormete, muestran la distribucion completa.

En cambio, las barras de error resumen la distribucion de los datos en un número como la desviacion estándar de los valores.

Hay al menos dos formas diferentes de mostrar las barras de error.

Aquí, agregamos la barra de error como argumento a un gráfico de barras. El argumento de la palabra clave `yerr` toma un número adicional, y lo muestra como un marcador vertical adicional.

```python
fig, ax  = plt.subplots()

ax.bar('Rowing',
      summer2016_nadadores['Height'].mean(),
      yerr = summer2016_nadadores['Height'].std())

ax.bar('Gymnastics',
      mens_gymnastics['Height'].mean(),
      yerr = mens_gymnastics['Height'].std())

ax.set_ylabel('Height (cm)')

plt.show()
```

Es útil por que resume la distribucion completa que vio en los histogramas de dos números: el valor medio y la dispersión de valores, cuantificados como la desviación estándar.


## Agregar barras de error a los grafica de barras: Metodo 2

También podemos agregar barras de error a un gráfico de líneas.

Para graficar estos datos con barras de error, usaremos el método de barras de error Axes (`errorbar`).

Al igual que el método de trazado, este método toma una secuencia de valores x, en este caso, la columna "MES", y una secuencia de valores de y, en este caso, la columna con las temperaturas mensuales media normales. Además, un argumento de palabra clave `yerr` puede tomar la columna en los datos que contiene las desviaciones estándar de las temperaturas medias mensuales.


```python
fix, ax = plt.subplots()

ax.errorbar(seattle_weather['DATE'],
           seattle_weather['MLY-TAVG-NORMAL'],
           yerr = seattle_weather['MLY-TAVG-STDDEV'])

ax.errorbar(austin_weather['DATE'],
           austin_weather['MLY-TAVG-NORMAL'],
           yerr = austin_weather['MLY-TAVG-STDDEV'])

ax.set_ylabel('Temperature (Fahrenheit)')

plt.show()
```

Al igual que antes, esto agrega marcadores verticales a la trama, que se ven así.

## Personalizar los Diagrama de Dispersión

Podemos personalizar los diagramas de dispersión de una manera simular a la personalización que introdujimos en otros diagramas.

Por ejemplo: si queremos mostrar dos comparaciones bivariadas uno al lado del otro, queremos asegurarnos de que sean visualmente distintos.

Aqui, vamos a trazar dos diagramas de dispersión en los mismo ejes. En uno, mostraremos los datos de los años ochenta y en el otro, mostraremos los datso de la cécada de los noventa.

Podemos seleccionar estas partes de los datos, usando la indexación de serie de tiempo que ha visto antes para crear dos DataFrames llamados ochenta y noventa. Luego, agregamos cada uno de estos DataFrames en el objeto Axes.

Primero, sumamos los datos de los años ochenta. Agregamos personalización: Establecemos el color de los punto en rojo y etiquetamos estos datos con la cadena "ochenta". Luego, sumamos los datso de los noventa. Estos puntos serán azules y los etiquetaremos con la cadena "noventa".

Llamamos al método de leyenda para agregar una letenda que nos dirá qué DataFrame se identifica con qué color, agregamos las etiquetas del eje y llamamos plt.show().

```python
eigthties = clima['1980-01-01':'1989-12-31']
nineties = clima['1990-01-01':'1999-12-31']

fix, ax =plt.subplots()

ax.scatter(eigthties['co2'], eigthties['relative_temp'],
          color = 'red', label = 'eighties')
ax.scatter(nineties['co2'], nineties['relative_temp'],
          color = 'blue', label = 'nineties')

ax.legend()

ax.set_xlabel('CO2 (ppm)')
ax.set_ylabel('Relative temperature (Celsius)')

plt.show()
```

## Diferentes formatos de archivo

**.png**: Este formato de archivo proporciona una compresión sin pérdidas de su imagen. Eso significa que laimagen conservará una alta calidad, pero también ocupan cantidades relativamente grandes de espacio en disco o ancho de banda.

**.jpg**: Este formato utiliza compresión con pérdida y se puede utilizar para crear figuras que ocupen menos espacio en disco y menos ancho de banda.

Puede controlar qué tan pequeño será el archivo resultante y el grado de pérdida de caldiad, estableciendo el argumento de la palabra clave `quality` (`fig.savefig(nombre.extension, quality = 50)`). Este será un número entre 1 y 100, pero debe evitar los valores por encima de 95, porque en ese punto la compresión ya no es efectiva.

**.svg**: La elección del formato de archivo svg producirá un archivo de gráficos vectoriales donde diferentes elementos se pueden editar en detalles miediante software de gráficos avanzado, como Gimp o Adobe Illustrator.

Si necesita editar la figura despues de producirla, esta podría ser una buena opción.


## Resolucion

Otra palabra calve que puede utilizar para controlar la calidad de las imágenes que produce es el argumento de la palabra clave `dpi` (`fig.savefig(nombre.extension,dpi = 300)`). Esto significa puntos por pulgada, cuando mayor sea este número, más densamente se renderizará la imagen.

Si establece este número en 300, por ejemplo, esto generará una resolución de bastante alta calidad de su imagen en el archivo. Por supuesto, cuanto mayor sea la resolución que solicite, mayor será el tamaño del archivo.

Un columna enun marco de datos de pandas es un objeto de serie de pandas, por lo que podemos obtener la lista de diferentes deportes presentes en los datos llamados al método único de esa columna (`unique()`).

```python
sports = summer2016['Sport'].unique()
print(sports)
```

## Diagrama de barra de alturas de todos los deportes

Digamos que nos gustaria visualizar la altura de atletas en cada uno de los deportes, con una barra de error de desviacion estándar. Dado que no sabemos de antemano cuántos deportes hay en el DataFrame, una vez que hemos extrado los valores únicos, podemos recorrerlos.

En cada iteración, configuramos una variable de bucle llamada sport para que sea igual a uno de estos valores únicos.

Luego creamos un DataFrame más pequeño, al que llamamos "sport_df", seleccionando las filas en las que la columna "Deporte" es igual al deporte seleccionado en esta iteración.

Podemos llemar al método de barras de los ejes que creamos para esta gráfica. como antes, se llama con la cadena que lleva el nombre del deporte como primer argumento, el método medio de la columna "Altura" se establece en la altura de la barra y se establece una barra de error para que sea igual a la desviacion estandar de valor en la columna.

Despues de iterar sobre todos los deportes, salimos del ciclo. Luego podemos configurar la eituqeta y para indicar el significado de la altura de cada barra y podemos configurar las etiquetas de marca del eje X para que sean iguales a los nombres de los deportes. Como hicimos con los nombres de los países en el gráfico de barra apiladas que vio en la leccion anterior, giramos estas etiquetas 90 grados, para que no se atropellen entre sí.

```python
fig ,ax = plt.subplots()
fig.set_size_inches([11,5])
for sport in sports:
    sport_df = summer2016[summer2016['Sport'] == sport]
    
    ax.bar(sport, sport_df['Height'].mean(),
          yerr = sport_df['Height'].std())
    
ax.set_ylabel('Height (cm)')
ax.set_xticklabels(sports, rotation = 90)
plt.show()
```

# Introducción a las tramas y subtramas relacionales


Muchas preguntas en ciencia de datos se centran en describir la relación entre dos variables cuantitativas, Seaborn llama tramas que visualizan esta relación "tramas relacionales". anteriormente haciamos comparacion entre subgrupos con el parametro `hue`, y se graficaba en la misma parcerla, ahora haremos lo mismo, pero en diferentes graficos con una nueva funcion llamada `relplot()`.

## Introducción a `relplot()`

`relplot()` significa "gráfico relacional" y le permite visualizar la relación entre dos variables cuantitativas usando diagramas de dispersión o diagramas de líneas.

El uso de `relplot()` nos brinda una gran ventaja: la capacidad de crear subtramas en una sola figura.

## scatterplot()  vs. relplot()

Usando `scatterplot()`


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x = 'total_bill',
               y = 'tip',
               data = tips)
plt.show()
```

Usando `relplot()`

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.relplot(x = 'total_bill',
            y = 'tip',
            data = tips,
           kind = 'scatter')#kind:Especificar el tipo de diagrama relacional a usar->Diagrama de dispersión o diagrama de lineas

plt.show()
```

# Personalizacion de diagrama de dispersión

## Subgrupos con tamaño de puntos

Crearemos un gráfico de dispersión de la factura total frente al monto de la propina

Queremos que cada punto en el diagrama de dispersión sea de tamaño basado en el número de personas en el grupo, con grupos más grandes que tienen puntos más grandes en la trama.

Para que logremos interpretar de una mejor manera el grafico (debido a que por defecto todos los puntos son del mismo color) es posible cambiar el color de los puntos dependiendo de su tamaño, para ello se usa el parametro `hue` igual al nombre de la varibale "size".

# Diagrama de caja


Un diagrama de caja muestra la distribución de los datos cuantitativos.

El cuadro de color representa el percentil 25 al 75, y la linea en el medio del cuadro representa la mediana. Los bigotes dan una idea de la dispersión de la distribución y los punto flotantes representan valores atípicos.

Los diagrama de caja se usan comúnmente como una forma de comparar la distribucion de una variable cuantitativa a través de diferentes grupos de una variable categórica.

Ejemplo:

El diagrama de caja que se muestra aqui usa el conjunto de datos de consejos y compara la distribución de la factura pagada por mesa entre los diferentes diías de la semana.

A partir de este diagrama de caja podemos ver rápidamente que la factura mediana es más alta los sabados y domingos, pero la dispersión de la distribución tambien es mayor.

Esta comparación sería mucho más dificil de hacer con otros tipos de visualizaciones. 

<img src=".\img\15.png" width="350px" height="300px">


## Como crear un diagrama de caja

Si bien Seaborn tiene una función `boxplot()`, usaremos la funcion `catplot()`, porque facilita la creaci´n de subparcelas usando los parámetros `col` y `row`.

```python
g = sns.catplot(x = 'time',
                y = 'total_bill',
                data = tips,
                kind = 'box')

plt.show()
```

## Cambiar el orden de las categórias

Como recordatorio, `catplot()` le permite cambiar el orden de las categorías usando el parámetro `order`.

```python
g = sns.catplot(x = 'time',
                y = 'total_bill',
                data = tips,
                kind = 'box',
               order = ['Dinner',
                       'Lunch'])

plt.show()
```

## Omitir los valores atipicos del diagrama


A veces es posible que no necesitemos ver los valores atipicos en los diagramas, y para ello hay un paramatro para desactivarlos, y lo hacemos con el parametro `sym = ""`.

Si paso una cadena vacia a `sym`, omitirá por completo los valores atípicos de su gráfico.

`sym` también se puede usar para cambiar la apariencia de los valores atípicos en lugar de omitirlos.

```python
g = sns.catplot(x = 'time',
                y = 'total_bill',
                data = tips,
                kind = 'box',
               sym = "")

plt.show()
```

## Cambiar los bigotes usando `whis`

De forma predeterminada, los bigotes se extienden hasta 1.5 veces el rango intercuatílico o "IQR".

El IQR es el percentil 25 al 75 de una distribucion de datos.

Si desea cambiar la forma en que se efinen los bigotes en su diagrama de caja, puede hacerlo usando el parámetro `whis`.

Hay varias opciones para cambiar los bigotes. Puedes cambiar el rango de lo bigotes desde 1.5 veces el IQR (que es el predeterminado) a 2 veces el IQR configurando --> `whis = 2.0`.

Alternativamente, puede hacer que los bigotes definan puntos inferiores y percentiles superiores pasando una lista de los valores inferior y superior.

En este ejemplo, pasar "[5, 95]" dará como resultado que el bigote inferior sea dibujado en el percentil 5 y el bigote superior se dibuja en el percentil 95.  -->  `whis = [5, 95]`.


Finalmente, es posible que solo desee dibujar los bogotes en los valores mínimo y máximo. Puede hacerlo especificando el percentil inferior como 0 y el percentil superior como 100: --> `whis = [0, 100]`.


Aqui un ejemplo donde los bogotes se establecen en los valores mínimo y máximo.

Tenga en cuenta que no hay valores atípicos, porque la caja y los bigotes cubren todo el rango de datos.

```python
g = sns.catplot(x = 'time',
                y = 'total_bill',
                data = tips,
                kind = 'box',
               whis = [0, 100])

plt.show()
```

# Diagrama de puntos: Grafica categórica


Los diagramas de puntos muestran la media de una variable cuantitativa para las observaciones en cada categória, trazadas como un solo punto. Esta gráfica de puntos utiliza el conjunto de datos de propina y muestra la factura promedio entre fumadores y no fumadores.

Las barras verticales que se extienden por encima y por debajo de la media representan los intervalos de confianza del 95% para esa media. Al igual que los intervalos de confianza que vimos en los diagrama de líneas y de barras, los invervalos de confianza nos muestran el nivel de incetidumbre que tenemos sobre estas estimaciones medias.

Suponiendo que nuestros datos son una muestra aleatoria de una poblacion, podemos estar 95% seguros de que la verdadera media de la población en cada grupo se encuentra dentro del intervalo de confianza que se muestra.



## Diagrama de lineas vs Diagramas de puntos

Tanto los gráficos de líneas como los gráficos de puntos muestran la media de una variable cuantitativa y los intervalos de confianza del 95% para la media. Sin embargo, hay una diferencia clave.

En una gráfica de puntos, un eje, generalemten el eje X, es ima variable categórica, lo que la convierte en una gráfica categórica.


También puede estar pensando: los diagramas de puntos parecen mostrar la misma información que los diagramas de barras. Para cada categória, ambos muestran la media de una variable cuantitativa y los intervalos de confianza para esas medias.

## Mostrando la mediana


Este es un gráfico de punto de la factura promedio que compara a los fumadores con los no fumadores. Para que los puntos y los intervalos de confianza se calculen para la mediana en lugar de la media, importe la función mediana de la biblioteca `numpy` y establezca `estimator = median`.

```python
import matplotlib.pyplot as plt  
import seaborn as sns
from numpy import median

sns.catplot(x = 'smoker',
           y = 'total_bill',
           data = tips,
           kind = 'point',
           estimator = median)

plt.show()
```

La mediana es más resistente a los valores atípicos, por lo que si su conjunto de datos tiene muchos valores atípicos, la mediana puede ser una mejor estadística para usar.

## Objetos: FacetGrid vs. AxesSubplot

Las funciones de trazado de Seaborn crean dos tipos diferentes de objetos: FacetGrids y AxesSubplots. 

Para averiguar con qué tipo de objeto está trabajando, primero asigne el resultado de la gráfica a una variable. En la documentación, la variabla menudo se denomina **"g"**, por lo que también lo haremos aqui. Escriba `type(g)` para devolver el tipo de objeto.

**`g = sns.scatterplot(x = 'height', y = 'weight', data = df)`**

**`type(g)`**  # -> Me dira si se trata de un objeto FacetGrid o AxesSubplot.



Recuerde que `relplot()` y `catplot()` admiten la creacion de subtramas. Esto significa que están creando objetos FacetGrid.

Por el contrario, las funciones gráficas de tipo único como `scatterplot()` y `countplot()` devuelve un solo objeto AxesSubplot.

## Agregar un titulo a FacetGrid

Recuerde que `catplot()` habilita las subtramas, por lo que devuelve un objeto FacetGrid.

Para agregar un título a un objeto FacetGrid asigne el gráfico a la variable `g`. Después de asignar la trama a `g`, puede configurar el título usando  `g.fig.suptitle("Titulo")`



```python
sns.set_context('talk')

g = sns.catplot(x = 'smoker',
           y = 'total_bill',
           data = tips,
           kind = 'point',
           capsize = 0.2)

g.fig.suptitle('Titulo de Prueba')

plt.show()
```

## Ajustar la altura del titulo en un FacetGrid


Tenga en cuenta que, de forma predeterminada, el titulo de la figura puede ser un poco bajo. Para ajustar la altura del título, puede usar el parámetro `y`, el valor predeterminado es `1.0`, por lo que configurando en  `1.03` lo hara un poco más alto que el valor predeterminado.



```python
sns.set_context('talk')

g = sns.catplot(x = 'smoker',
           y = 'total_bill',
           data = tips,
           kind = 'point',
           capsize = 0.2)

g.fig.suptitle('Titulo de Prueba',
               y = 1.03)

plt.show()
```

## Agregar un titulo a un AxesSubplot

Para asignarle un titulo a un objeto AxesSubplot, primero debemos asignar la grafica a una variable, normalmente se la asigna a una variable de nombre `g`, y luego para agregar el titulo se usa el metodo `g.set_title("Titulo", y = 1.03)`, el parámetro `y` es usado para modificar la altura del titulo.

```python
g = sns.countplot(x = 'smoker',
             data = tips,
             hue = 'sex')

g.set_title('Titulo de Prueba',
           y = 1.03)
plt.show()
```

## Titulos para Sub diagramas

Dado que `g` es un objeto FacetGrid, usar `g.fig.suptitle()` agregará un titulo a la figura como un todo.

Para modificar los títulos de las subparcelas, utilize `g.set_titles` para establecer los títulos de cada AxesSubplot.

Si desea usar el nombre de la variable en el título, puede usar el `{col_name}` para hacer referencia al valor de la columna.

```python
g = sns.relplot(x = 'total_bill',
            y = 'tip',
            data = tips,
           kind = 'scatter',
            col = 'day',
           col_wrap = 2)

g.fig.suptitle('Titulo General', y = 1.03)

g.set_titles('Day: {col_name}')

plt.show()
```

## Agregar etiquetas de eje para FacetGrid y AxesSubplot 

Para agregar etiquetas de eje, asigne la gráfica a una variable y luego llame a la función `g.set(xlabel = 'etiqueta_X', ylabel = 'etiqueta_Y')`.

```python
sns.set_context('talk')

g = sns.catplot(x = 'smoker',
           y = 'total_bill',
           data = tips,
           kind = 'point',
           capsize = 0.2)

g.fig.suptitle('Titulo de Prueba',
               y = 1.03)

g.set(xlabel = 'Eje X', ylabel = 'Eje Y')

plt.show()
```

En cambio, después de crear la trama, llamamos a la función de `matplotlib`  `plt.xticks(rotation = 90)`, que rota las etiquetas de marca 90 grados. Esto funciona con los objetos FacetGrid y AxesSubplots.

# Personalizando con matplotlib

Seaborn tiene varias funciones convenientes para modificar componentes de su visualización final de Seaborn. Dado que seaborn se basa en Matplotlib, existe una ampli variedad de opciones para modificar aún más sus gráficos de Seaborn.


## Matplotlib Ejes

Al usar los objetos de ejes de matplotlib, puede personalizar casi cualquier elemento de su gráfico. 

El concepto más importante es agregar código adicional para crear las subtramas usando las funciones de subparcelas de matplotlib y pase el objeto de ejes resultante a la función Seaborn. Seaborn luego trazará los datos en los ejes dados.

```python
fig, ax = plt.subplots()
sns.histplot(df.Tuition, ax = ax)
ax.set(xlabel = 'Tuition 2013-14')
plt.show()
```

## Otras personalizaciones

Este ejemplo muestra cuánto control tiene una vez que comienza a modificar los ejes. Además de configurar una etiqueta x, también se puede cambiar la etiqueta Y, y establecer un limite para que los rangos de matriculas suan solo entre 0 y $50000.

```python
fig, ax = plt.subplots()
sns.histplot(df.Tuition, ax = ax)
ax.set(xlabel = 'Tuition 2013-14',
       ylabel = 'Distribution', xlim=(0,50000),
       title = '2013-2014 Tuition and Fess Distribution')
plt.show()
```

## Combinar gráficos

Este ejemplo ilustra cómo puede combinar y personalizar varias parcelas de Seaborn.

Primero, creamos dos ejes y nos aseguramos de que el eje Y se comparta entre las parcelas. Luego trazamos un histograma completo en ax0. Luego, usamos la función de consulta de pandas para trazar solo los datos del estado de Minnesota.

El otro concepto nuevo es el uso de la línea `axv` para indicar la cantidad máxima que podemos presupuestar para la matrícula, en este caso, $20000.

El paso final es activar la leyenda y mostrarla.



```python
fig, (ax0, ax1) = plt.subplots(nrows = 1, ncols = 2,
                               sharey = True, figsize = (7, 4))

sns.histplot(df.Tuition, stat = 'density', ax = ax0)
sns.histplot(df.query('State == "MN"')['Tuition'], stat = 'density', ax = ax1)

ax1.set(xlabel = 'Tuition (MN)', xlim = (0, 70000))
ax1.axvline(x = 20000, label = 'My Budget', linestyle = '--')
ax1.legend()

plt.show()
```

## Gráficas Matriciales


El mapa de calor es el tipo más común de gráfico de matriz y Seaborn puede crearlo fácilmente. Estos tipos de gráficos matriciales pueden ser útiles para ver rápidamente las tendencias en un conjunto de datos.


## Obtención de datos en el formato adecuado

La función `heatmap()` de Seaborn espera que los datos estén en una matriz. Para ilustrar este proceso, podemos observar los datos de bicicletas para ver cómo varían los alquileres según el dia de la semana y el mes.

La función `crosstab()` crea una tabla para resumir los datos por día y mes. En este ejemplo, también usamos el argumento `aggfunc` para obtener el promedio de alquileres para cada una de las combinaciones de dia y mes.

La visualización del mapa de calor se puede personalizar de varias maneras para presentar la mayor cantidad de información con la mayor claridad posible.

# Usando FacetGrid, catplot y lmplot


Una de las características más poderosas de Seaborn es su capacidad para combinar pequeños multiples parcelas en una visualización más grande que puede ayudar a identificar tendencias en datos con muchas variables.


El concepto de pequeños múltiplos es útil para analizar datos con muchas variables.

La idea es que puede identificar rápidamente las tendencias en los datos, comparar múltiples parcelas una al lado de la otra usando las mismas escalas y ejes.

Estas parcelas se conocen como "parcela enrejadas" o "enrejedas".

En la ciencia de datos, este concepto también se denomica "facetas".

<img src=".\img\62.png" width="900px" height="300px">


En este ejemplo específico, podemos ver los datos de matrícula universitaria y cómo varía según el tipo de grado que ofrece la escuela, la region y si la escuela está controlada como una institución pública o privada.


## Datos ordenados

Un requisito muy importante para que Seaborn cree estos gráficos es que los datos deben estar en un formato ordenado.

Esto significa que cada fila de datos es una sola observación y las columnas contienen las variables.


<img src=".\img\63.png" width="700px" height="300px">

Una vez que los datos están en este formato, Seaborn puede realizar gran parte del trabajo pesado necesario para crear estas pequeñas parcelas múltiples.


## FacetGrid


El facetgrid es la base de muchas redes basadas en datos. 

Permite al usuario controlar cómo se distribuyen los datos en columnas, filas y tonas, y luego se utiliza para crear el tipo de gráfico adecuado.


El punto clave es recordar es que `FacetGrid()` proporciona mucha flexibilidad, pero debe usar un proceso de dos pasos para definir las facetas y mapear el tipo de gráfico.


## Ejemplo FaetGrid categórico 

Este ejemplo muestra cómo mapear un diagrama de caja en un `FacetGrid()` con reconocimiento de datos.

# Usando PairGrid y pairplot

El siguiente tipo de gráficos de cuadricula con reconcimiento de datos que analizaremos son `PairGrid` y `pairplot`.

Esto gráficos son similares a `FacetGrid`, `catplot` e `lmplot` porque nos permiten ver las interacciones entre diferentes columnas de datos. 

La diferencia con estos gráficos es que solo definimos las columnas de datos que queremos comparar.


## Relaciones de pareja

Echemos un vistazo a nuestros datos de viviendas justa de mercado que se vio en el ejercicio anterior.

El gráfico `PairGrid` nos permite construir gráficos que muestran las relaciones entre dos elemento de datos.

En este ejemplo, queremos ver la relaciones que podrían existir entre el alquiler justo de mercado y el ingreso medio de los inquilinos.

<img src=".\img\69.png" width="400px" height="300px">

Las diagonales contienen histogramas de Renta Justa de Mercado e Ingreso Medio.

La otra diagonal contiene un diagrama de dispersión que alterna qué variable está en el eje X o Y.


## Crear un `PairGrid`


El proceso para crear un `PairGrid` es similar a un `FacetGrid` en que debemos crear la cuadrícula, luego asignar un tipo de gráfico a la cuadrícula.

**`g = sns.PairGrid(df, vars = ['Fair_Mrkt_Rent', 'Median_Income'])`**

**`g = g.map(sns.scatterplot)`**

<img src=".\img\70.png" width="300px" height="300px">

El punto interesante de esta cuadrícula es que no definimos los parámetros de fila y columna. En su lugar, definimos las variables.

En este caso, las variables son las columna del marco de datos que queremos ver.

El marco de datos contiene una columna de Renta Justo de Mercado, así como una columna de Ingreso medio, que pasamos a `PairGrid`.

El siguiente paso es hacerle saber a Seanborn que queremos ver una visualización de diagrama de dispersión de la relación entre las dos variables.

Seabor se encarga de variar los ejes X e Y para cada uno de estos gráficos.


## Personalizar las diagonales de el `PairGrid`


En el ejemplo anterior, las diagonales contenian líneas rectas, que no son muy perpicaces.

Afortunadamente, `PairGrid` admite la definición del tipo de gráficos que se pueden mostrar en las diagonales.

Puede utilizar la función `map_diag` para definir la función de trazado de la diagonal principal.

La función `map_offdiag` define la otra diagonal.


**`g = sns.PairGrid(df, vars = ['Fair_Mrkt_Rent', 'Median_Income'])`**

**`g = g.map_diag(sns.histplot)`**

**`g = g.map_offdiag(sns.scatterplot)`**

<img src=".\img\71.png" width="300px" height="300px">

En este caso, estamos viendo un gráfico de histograma de la Renta Justa de Mercado y la Renta Media. Esta vista es mucho más útil que la dispersión anterior.


## Pairplot

Pairplot es una función de atajo que simplifica el `PairGrid`.

Primero, definimos las variables como lo hicimos anteriormente. Sin embargo, en lugar de mapear las parcelas, podemos pasar palabras clave para especificar el tipo de gráfico para las diagonales principales y secundarias.

El otro buen aspecto de esta función es que traza una línea de regresión para nosotros.


**`sns.pairplot(df, vars = ['Fair_Mrkt_Rent', 'Median_Income'], kind = 'reg',
                diag_kind = 'hist')`**
                
<img src=".\img\72.png" width="350px" height="300px">                


Esta es una función útil para visualizar posibles relaciones en los datos.


## Personalizar un `pairplot()`

Aquí hay otro ejemplo de cómo Seaborn puede generar tramas complejas con muy poco código.

Para limitar los datos que nos interesan, filtramos los datos para alquileres con menos de 3 dormitorios y asigne el matiz a la columna de dormitorios. **`sns.pairplot(df.query('BEDRMS < 3'),

o `diag_kws={'alpha':.5}`

# Uso de JointGrid y jointplot

La última cuadrícula con reconocimiento de datos que analizaremos es `JointGrid` y su función complementaria, `jointplot()`.

Un `JointGrid()` nos permite comparar la distribución de datos entre dos variables.

Un `JointGrid()` hace uso de diagramas de dispersión, regresión, líneas, así como histogramas, diagramas de distribución y estimaciones de densidad de kernel para darnos una idea de nuestros datos.


## Visión general de `JointGrid()`

El `JointGrid()` es similar a los otros gráficos Grid en que toma una pequeña cantidad de entradas y crea una visualización perspicas de los datos.


La entrada de `JointGrid()` es una variable X e Y. En este caso, estamos analizando la relación entre la matrícula universitaria y la tasa de admisión.

El centro del gráfico contiene un gráfico de dispersión de esta dos variables.

Los gráficos a lo largo de los eje X e Y muestran la distribución de los datos para cada variable.

<img src=".\img\75.png" width="300px" height="300px">

Esta parcela se puede configurar especificando el tipo de parcelas conjuntoas así como la parcela marginales.



## `JointGrid()` Basico

El procesi de creación de `JointGrid()` sigue los mismo pasos que las otras cuadrículas.

Debe definir la cuadrícula y mapear las parcelas en la cuadrícula.

En este caso, definimos las tasa de matrículas y admisión como las variables X e Y.

Luego mapeamos `regplot()` y `histplot()` de Seaborn a la cuadrícula. Seaborn se encarga de crear la trama.

**`g = sns.JointGrid(data = df, x = 'Tuition', y = 'ADM_RATE_ALL')`**

**`g.plot(sns.regplot, sns.histplot)`**

<img src=".\img\75.png" width="300px" height="300px">




## `JointGrid()` Avanzado

Aquí mostramos la flexibilidad disponible a través de la función `JointGrid()`.

La función `plot_joint()` especifica que se debe incluir un gráfico `kde` en el centro.

Los `kdeplots` en los márgenes se defininen con `plot_marginals()`.

Esta vista de los datos nos ayuda a comprender las distribuciónes de datos para estas dos variables.

**`g = sns.JointGrid(data = df, x = 'Tuition', y = 'ADM_RATE_ALL')`**

**`g = g.plot_joint(sns.kdeplot)`**

**`g = g.plot_marginals(sns.kdeplot, shade = True)`**

<img src=".\img\76.png" width="300px" height="300px">

En este ejemplo, hay un gran grupo de valores que tienen matrícula entre $15000 y $20000, y tasa de admisión de alrededor del 60%.



## `jointplot()`

El `jointplot()` es más fácil de usar pero proporciona menos capacidades de personalización.

En este ejemplo, mostramos un gráfico hexadecimal como método para demostrar la relación entre las dos variables.

Como puede ver, este es un enfoque más simple que crear `JointGrid()`.


**`sns.jointplot(data = df, x = 'Tuition', y = 'ADM_RATE_ALL', kind = 'hex')`**

<img src=".\img\77.png" width="300px" height="300px">


## Personalizar `jointplot()`

El `jointplot()` admite la creación simple de diagramas de dispesión, hexadecimales, residuales, de regresión y kde. También puede admitir la adición de gráficos superpuestos para mejorar el resultado final.


Este ejemplo muestra un paradigma bastante común para analizar los datos de Seaborn.

Ademas de definir las variables X e Y, establecemos los límites para el corte del eje x en 0 y 25000.

También pasamos palabras clave al gráfico marginal para controlar la estructura del histograma.

En este caso, solo queremos ver la distribución de resultados para universidades públicas con una matrícula de pregrado inferior a 2500 estudiantes.

Finalmente, al incluir la función `plot_joint()`, se superpone un diagrama de kde en el diagrama de dispersión.

**`g = (sns.jointplot(x = 'Tuition',
                     y = 'ADM_RATE_ALL',
                     kind = 'scatter',
                     xlim = (0, 25000),
                     data = df.query('UG < 2500 & Ownership == "Public"')).plot_joint(sns.kdeplot))`**

El diagrama de `scatter()` de matplotlib es un método muy simple para comparar la interacción de dos variables en el eje X e Y.

El `lmplot()` combina muchas de estas características del `regplot()` subyacente y `residplot()` además de la capacidad de trazar los datos en un FacetGrid().

En mucho casos, `lmplot()` es la mejor función para determinar relaciones lineales entre datos.
