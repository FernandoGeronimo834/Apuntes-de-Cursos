* **`git pull <nombre de rama remoto (origin)> <nombre de la rama donde quiero traer las actualizaciones (ej: main)>`**: De esta forma me traigo las ultimas actualizaciones del repositorio remoto a mi repo local, esto deberia hacerlo siempre antes de enviar un cambio al repo remoto, ya que puede suceder que mientras yo trabajaba en mi repo local, otro desarrollador ya mando un una actualizacion al repositorio remoto, por lo cual si no hago esto antes de actualizar mi commit en el repo remoto podria saltear las actualizaciones de otro desarrollador.

## Teoria:


* **Directorio**: Carpeta donde se encuentra nuestro proyecto, y los archivos de nuestro proyecto.
* **`git init`**: Al usar este comando pasan muchas cosas, en particular suceden 2, una es que se crea un espacio en memoria ran que se llama `Staging`(Es un area aislada en memoria ram donde en principio ire agregando los cambios del proyecto), y la segunda cosa que sucede al usar `git init` es que se creara un repositorio (donde dentro de la carpeta del proyecto o directorio, se creara una carpeta `/.git/`, donde en esa carpeta estaran todos los cambios al final de nuestro proyecto).
* **`git add nombre_de_archivo.extension`**: Al usar este comando, cuando quiero guardar cambio a uno o varios archivos nuevos o modificados, estos primero se guardaran en el `Staging`, donde estos archivos estaran esperando a que lo subamos al repositorio.

* **Observacion de `git add ... `**: Antes de usar el comando `git add ...` a un archivo, el mismo se clasifica como `"untracked"` es decir que aun git no esta detectando si se realizan cambios sobre dicho archivo. Una vez que a un archivo se le aplica el comando `git add ...`, `git` estara atento a todos los cambios que se realizen sobre dicho archivo, donde ademas el archivo se clasifica como `tracked` y el mismo es subido a la zona de `Staging`.

* **`git rm --cached nombre_de_archivo.extension`**: Puedo bajarlo archivos que ya se encuentren en el sector de `Staging`.
* **`git commit -m "Acotacion sobre el cambio a guardar"`**: Una vez que los archivos se encuentren en la zona de `Staging` puedo subirlos al repositorio con este comando. Donde este repositorio donde se guardaran los cambios de llamara `MASTER`, y es aqui donde estaran todos los cambios que realize y subi al repositorio.
* **Observacion del `git commit ...`**: Cada ver que realize `git commit ... ` sobre archivos que ya se encontraban en el repositorio, en dicho repositorio se estaran creando diferentes versiones del archivo (version_1, version_2, version_3, ..., version_n). Donde el numero raro del commit (caracteristica del cambio), es el identificador de dicho cambio en la base de datos.
#### Para traer un cambio que esta en el repositorio, pero no esta en mi carpeta, por ejemplo: si un miebro del equipo realizo un cambio en el archivo, el cual yo no lo tengo en mis archivos locales, ya que trabajamos sobre el mismo archivo.

* **ir a rama: `MASTER`**: Debo ir a la rama `MASTER`
* **`checkout`**: Traer los ultimos cambios o los cambios que yo quiero a mi carpeta (Pueden ser todos los cambios, los cambios de solo algunos archivos o cambios de solo un archivo).


## Ramas:

* `MASTER`    ---`Version_1`----`Version_2`----...----`Version_Actual`

* `Experimetos`, normalmente a la rama de Experimetos de le llama `development`: Donde copio una version de la rama `MASTER`, en el que puedo realizar experimientos y realizar `commits` en la rama `Experimetos` y por cada `Commit` se creara una nueva version de la copia que realize de la rama `MASTER` pero sin afectar la versiones de la rama `MASTER` 

* `Bugfixing` o `Hotfix`: Cuando encuentro un bug en la `Version_Actual` de la rama `Master`, realizo una copia de la version actual de la rama `Master` en esta rama (`Hotfix`), donde ire realizando cambios con el fin de solucionar este bug. Estos cambios de reparacion de Bugs deben resolverse lo mas pronto posible, para no acarrear el error en las siguientes ramas de `develoment`.

* **`merge`**: Una vez que termine de solucionar un bug o deseo unir los cambios de una rama con otra rama, debo usar el comando `merge`.

# Fusion de Ramas:
* Tener en cuenta que al usar el comando `merge` debo hacerlo donde quiero que se lleven a cabo todas las actualizaciones. Es decir que si me encuentro en el `HEAD` apuntando al `master` y uso el comando `merge`los cambios realizado en la otra rama se fusionaran a la rama `master` y continuara el flujo normal como cuando trabaja con la rama `master`. Por eso es importante aplicar el comando `merge` en la rama donde quiero que vengas las actualizaciones.

* Al aplicar `merge` en la rama `master`, se fusionara el ultimo commit de la rama `master`, con el ultimo commit de la rama que deseo fusionar.
* Es posible que aparescan conflictos, esto sucede cuando una misma linea es modificada en ambas ramas, y estas intentan fusionarse.

* **`git merge nombre_de_la_rama `**: Me abrira otra ventana donde debere colocarle el nombre del commit de la fusion. El nombre de la rama sera, la rama que deseo fusionar con la rama `master`. Lo que sucedera al usar este comando es crear un nuevo commit en la rama `master`, y me notificara si encuentra algun conflicto.

* Observacion: Si ya fusione las ramas, pero quiero seguir realizando cambios en la rama de experimentos, pero quiero traer la ultima actualizacion del commit de master, debo usar el comando de `merge` dentro de la rama de experimentos `git merge master`, y luego tendre el ultimo commit de master en mi rama de experimentos.

* **Conflictos**: **(Nombre_de_rama|MERGING)** Cuando entre en conflictos al momento de unir las ramas,debo arreglarlo manualmente, donde debo abrir los archivos y señalar con cual de las dos opciones donde se modifico la misma linea deseo quedarme, luego de seleccionar con cual me voy a quedar deberia borrarse la otra opcion (esto en VSC), luego debo guardar los cambio en el archivo, y por ultimo uso el comando `git commit -am "Descripcion referia a la fusion y solucion del conflicto"`.

# Diferencias entre comandos: (IMPORTANTE!)

Git reset y git rm son comandos con utilidades muy diferentes, pero se pueden confundir muy fácilmente.

### git rm
Este comando nos ayuda a eliminar archivos de Git sin eliminar su historial del sistema de versiones. Esto quiere decir que si necesitamos recuperar el archivo solo debemos “viajar en el tiempo” y recuperar el último commit antes de borrar el archivo en cuestión.

Recuerda que git rm no puede usarse así nomás. Debemos usar uno de los flags para indicarle a Git cómo eliminar los archivos que ya no necesitamos en la última versión del proyecto:

**`git rm --cached  nombre_de_archivo.extension `**: Elimina los archivos de nuestro repositorio local y del área de staging, pero los mantiene en nuestro disco duro. Básicamente le dice a Git que deje de trackear el historial de cambios de estos archivos, por lo que pasaran a un estado untracked.

**`git rm --force nombre_de_archivo.extension`**: Elimina los archivos de Git y del disco duro. Git siempre guarda todo, por lo que podemos acceder al registro de la existencia de los archivos, de modo que podremos recuperarlos si es necesario (pero debemos usar comandos más avanzados).

### **`git reset`**

Este comando nos ayuda a volver en el tiempo. Pero no como git checkout que nos deja ir, mirar, pasear y volver. Con git reset volvemos al pasado sin la posibilidad de volver al futuro. Borramos la historia y la debemos sobreescribir. No hay vuelta atrás.

Este comando es muy peligroso y debemos emplearlo solo en caso de emergencia. Recuerda que debemos usar alguna de estas dos opciones:

Hay dos formas de utilizar **`git reset`**: con el argumento --hard, borrando toda la información que tengamos en el área de staging (y perdiendo todo para siempre). O, un poco más seguro, con el argumento --soft, que mantiene allí los archivos del área de staging para que podamos aplicar nuestros últimos cambios pero desde un commit anterior.

**`git reset codigo_comit --soft`**: Borramos todo el historial y los registros de Git pero guardamos los cambios que tengamos en Staging, así podemos aplicar las últimas actualizaciones a un nuevo commit.

**`git reset codigo_comit --hard`**: Borra todo. Todo todito, absolutamente todo. Toda la información de los commits y del área de staging se borra del historial.
¡Pero todavía falta algo!

**`git reset HEAD nombre_de_archivo.extension`**: Este es el comando para sacar archivos del área de staging. No para borrarlos ni nada de eso, solo para que los últimos cambios de estos archivos no se envíen al último commit, a menos que cambiemos de opinión y los incluyamos de nuevo en staging con git add, por supuesto.
¿Por qué esto es importante?
Imagina el siguiente caso:

Hacemos cambios en los archivos de un proyecto para una nueva actualización. Todos los archivos con cambios se mueven al área de staging con el comando git add. Pero te das cuenta de que uno de esos archivos no está listo todavía. Actualizaste el archivo, pero ese cambio no debe ir en el próximo commit por ahora.

¿Qué podemos hacer?

Bueno, todos los cambios están en el área de Staging, incluido el archivo con los cambios que no están listos. Esto significa que debemos sacar ese archivo de Staging para poder hacer commit de todos los demás.

¡Al usar git rm lo que haremos será eliminar este archivo completamente de git! Todavía tendremos el historial de cambios de este archivo, con la eliminación del archivo como su última actualización. Recuerda que en este caso no buscábamos eliminar un archivo, solo dejarlo como estaba y actualizarlo después, no en este commit.

En cambio, si usamos git reset HEAD, lo único que haremos será mover estos cambios de Staging a Unstaged. Seguiremos teniendo los últimos cambios del archivo, el repositorio mantendrá el archivo (no con sus últimos cambios, pero sí con los últimos en los que hicimos commit) y no habremos perdido nada.

Conclusión: Lo mejor que puedes hacer para salvar tu puesto y evitar un incendio en tu trabajo es conocer muy bien la diferencia y los riesgos de todos los comandos de Git.

### Otro punto de vista

**`Git rm`**: Remover/Eliminar

**`Git rm --cached`**: Elimina los archivos que están en staging (caché) pero los deja en tu carpeta (disco duro).

**`Git rm --force`**: Elimina el archivo de git y de tu carpeta (disco duro).
_Los archivos se pueden recuperar pero con comando avanzados.

**`Git reset`**: Ir a archivos pasados.

**`Git resert --hard`**: Borra todo, del área de staging, commits e historial.

**`Git reset --soft`**: Borra el historial y los registros pero mantiene los cambios agregado a staging.

**`Git reset HEAD`**: Imagina que hiciste git add a dos archivos, un HTML y otro CSS. Sin embargo, te diste cuenta que no agregaste los estilos correctamente porque los titulos son rojos. Haces git reset HEAD, tu archivo se quita de staging, vuelve a tu carpeta, modificas el archivo css y vuelves a hacer git add. Si hubieses usado git rm tu archivo se hubiese eliminado.
Diferencias:

**`git rm`** solo elimina archivos (ya sea de staging o del disco duro).

**`git reset`** te envía a versiones antiguas y elimina archivos, historial y registros.

#  Flujo de Trabajos Profesionales.

Normalmente cuando se trabaja en un nuevo proyecto con un equipo de desarrolladores, la rama principal `main` se suele dirigir normalmente a un servidor de produccion final, donde el proyecto es puesto a prueba, este servidor puede ser un sitio web, donde se manda la version final como por ejemplo para luego subirse al play store. Pero cuando los desarrolladores terminan de trabajar en su respectivas ramas ya no hacen una fusion directa con la rama principal, sino que se intenta fusionar con una rama llamada `staging develop` donde se suelen  hacer las pruebas en un servidor de pruebas (este servidor tiene que ser lo mas parecido al servidor de produccion), cuando una rama se intenta fusionar con `staging develop` normalmente se aplica un `merge` y se fusionan, pero atravez de una funcion propia del `Github` llamada `pull request`, la cual antes de fusionar las ramas debe tener la aprobacion de los demas desarrolladores, quienes miraran tus archivos antes de fusionarlos, y daran el `okey` si les parece bien, luego de que todos los desarrolladores en el `okey`,`github` me dejara ejecuta el `merge`(esto puedo hacerlo yo o cualquier otro colaborador, llendo a `pull request` entrando al commit y hacer el merge), fusionando asi las ramas; cuando se observa que la fusiones de ramas funcionan de forma correcta en el `servidor de pruebas`, lo que prosigue es fusionar la rama del `staging develop` con la rama principal, lo cual al intentar hacer dicha funsion tambien pasa por el estado de `pull request`, donde todos los desarrolladores deben dar el `okey`, luego de darlo las ramas podran fusionarse y se podra ver el proyecto funcionando en el `servidor de produccion`.  Todo esto normalmente suele hacerlo una persona con el puesto de `DevOps`.

# Git cherry-pick  (Es una mala practica)

* Existe un mundo alternativo donde yo voy avanzado en una rama, pero necesito en master uno de esos avanzes de la rama, para ello se ocupa un comando llamado cherry-pick.

Ej:

* Modifico el archivo README.me , donde guardo los cambios, pero no lo voy a guardar para ejecutarlo en main, sino que lo voy a guardar para ejecutarlo en otra rama.

* Estando en `main` (donde modifique el README), aplico el `git stash`, luego estas modificaciones guardadas actualmente en el `stash` los voy a guardar en la otra rama con `git stash branch <nombre de la nueva rama donde se guardaran las modificaciones almacenadas en stash>`, donde luego de dar enter, estare dentro de la nueva rama creada, y alli es donde dare mi `git commit -am "<descripcion del commit>"`, si aplico `git status` puedo observar que mis cambios fueron guardados en mi nueva rama. Con `git log --oneline` podre ver todos los commit de las diferentes ramas creadas hasta el momento.

**Si ahora:**

El jefe quiere ver en la rama `main`, unos cambios que sucedieron (puede ser especificamente solo un par de lineas y no todo la modificaciones de la rama) en `<la nueva rama donde trabajo>`, pero aun no termine de trabajar con la rama como para hacer un merge, entonces:

* Estando en la rama `main` donde deseo traerme solo unas lineas (para este ejemplo) de la rama donde otro desarrollador aun se encuentra trabajando, lo que hago es usar el comando **`git cherry-pick <codigo commit>`**, como que hace este comando, no es realmente trarme lineas especificas a eleccion, sino que traigo modificaciones guardadas a partir de los diferentes commit creado en la rama del otro desarrollador, para ver el codigo del commit del otro desarrollador, primero debo estar en la rama del otro desarrollador con `git checkout <rama>`,y uso `git log --oneline`, para obtener el codigo de commit, que quiero usar para hacer **cherry-pick**, copio el codigo del commit (ej: 62b761e).

* Regreso a la rama principal con `git checkout main`, uso **`git cherry-pick <codigo commit>`**, le doy enter, y deberia traerme las modificaciones que se guardaron en el commit correspondiente al codigo que le pase al comando de `git cherry-pick <codigo commit>`, si le doy a `git status` no deberian verse los cambios, ya que `cherry-pick` pega las lineas inmediatamente. Pero si uso `git log --oneline` podre observar que a la rama `main` se agrego el commit que traje desde la rama del otro desarrollador.

* Si quiero enviar estos cambios el repositorio de Github, primero `git pull origin main` y luego `git push origin main` los cambios ya deberian aparecer en el repositorio de Github.

* Lo que hicimos anteriomente fue solo traernos un cambio que se guardo en un commit dentro de una rama paralela a la rama `main`, ahora cuando el desarrollador de la otra rama termina de trabajar con dicha rama y quiere implementar sus cambios a la rama `main`, debe hacer un `git merge <rama paralela a la principal>` (Esto debo ejecutarlo desde la rama a donde quiero implementar los cambios de la rama que tiene los cambios. En este caso ejecutare este comando desde `main`).

* Como al hacer `cherry-pick` ya habia copiado una linea de la nueva rama antes de que la nueva rama implemente los cambio en la rama principal, se va a generar un conflicto, el cual debemos resolver manualmente, guardar los cambios de archivo luego de resolver el conflicto, y luego hacer un `git commit -am "<descripcion>"` y enter.

* Para enviar la nueva implementacion a Github, primer hago `git pull origin main` y luego `git push origin main`. Luego en mi Github ya deberian aparecer los cambios, para Github nunca existio la rama `readme-mejorado`, solo sabe que mejore todo.

** Crear un comando de Git como un alias**

Si quiero ejecutar un comando por ej **`git shortlog -sn --all --no-merges`** y quisiera ejecutarlo como `git <alias>`, podria crear este ultimo comandod e la forma:

* **`git config --global alias.<alias> "<comando al que hara referencia el alias (sin el git inicial)>"`**: Por ejemplo `config --global alias.stats "shortlog -sn --all --no-merges"`, luego doy enter, y para ejecturar el comando `git shortlog -sn --all --no-merges`, podriamos hacerlo ahora como `git stats`

* **`git blame <archivo.extension>`**: Para ver linea por linea quien hizo que.
* **`git blame -c <archivo.extension>`**: Identa de una forma mas organizada el comando de arriba.

* **`git <comando> --help`**: Para saber como funciona un comando por dentro

* **`git blame < <nombre de carpeta>/<archivo>.<extension> o <archivo>.<extension> >  -L<numero1>,<numero2>`**: Para saber quien edito al ultimo las lineas entre la linea numero1 y numero2 (numero1<numero2). ej: `git blame perro.txt -L1,5`.

* **`git blame < <nombre de carpeta>/<archivo>.<extension> o <archivo>.<extension> >  -L<numero1>,<numero2> -c`**: Agrega un poco mas de formato, pero hace lo mismo que el comando de arriba.

* **`git branch`**: Ver ramas en repo local
* **`git branch -r`**: Ver ramas que estan en el repo remoto.

* **`git branch -a`**: Ver todas las ramas, tanto locales (en color blanco) , las remotas (en color rojo) y en verde en la rama en la que me encuentro ahora mismo.
